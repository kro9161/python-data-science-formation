## 📝 Partie 0 – Quel graphique pour quelle question ?

Avant de coder, la première question est toujours : **"Qu'est-ce que je veux montrer ?"**

| Question | Type de graphique | Fonction Matplotlib |
|----------|------------------|--------------------|
| Comment une valeur **évolue dans le temps** ? | Courbe (line) | `plt.plot()` |
| **Comparer** des catégories entre elles ? | Barres (bar) | `plt.bar()` / `plt.barh()` |
| Comment les valeurs se **distribuent** ? | Histogramme (hist) | `plt.hist()` |
| Y a-t-il une **relation** entre 2 variables numériques ? | Nuage de points (scatter) | `plt.scatter()` |
| Quelle est la **répartition en %** d'une catégorie ? | Camembert (pie) | `plt.pie()` |

> 💡 **Erreur classique :** utiliser un camembert pour comparer 8 catégories (illisible) ou une courbe pour des catégories sans ordre naturel. Le choix du graphique conditionne la lisibilité du message.

## 📝 Partie 1 – La courbe : évolution dans le temps

### 🔎 Quand l'utiliser ?
Quand les valeurs de l'axe X ont un **ordre naturel** (dates, jours, mois, étapes) et qu'on veut voir une **tendance** ou une **évolution**.

**Questions types :**
- *"Comment évoluent les ventes semaine par semaine ?"*
- *"La température monte-t-elle au fil des jours ?"*

```python
plt.plot(x, y)              # courbe simple
plt.plot(x, y1)             # première série
plt.plot(x, y2)             # deuxième série — sur le même graphique
```

````python
import matplotlib.pyplot as plt

# Ventes d'une boutique sur 6 mois
mois   = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun"]
pizzas = [120, 135, 160, 145, 180, 200]
burgers= [80,  90,  95, 100,  88,  110]

fig, ax = plt.subplots(figsize=(9, 4))

ax.plot(mois, pizzas,  color="steelblue", marker="o", linewidth=2, label="Pizzas")
ax.plot(mois, burgers, color="coral",     marker="s", linewidth=2, label="Burgers", linestyle="--")

ax.set_title("Évolution des ventes sur 6 mois")
ax.set_xlabel("Mois")
ax.set_ylabel("Quantité vendue")
ax.legend()
ax.grid(True, alpha=0.3)   # grille légère en arrière-plan
plt.tight_layout()
plt.show()
````

## 📝 Partie 2 – Les barres : comparer des catégories

### 🔎 Quand l'utiliser ?
Quand on veut **comparer des valeurs entre des catégories distinctes** qui n'ont pas d'ordre naturel entre elles.

**Questions types :**
- *"Quel produit se vend le mieux ?"*
- *"Quel département a le plus d'employés ?"*

```python
plt.bar(categories, valeurs)   # barres verticales
plt.barh(categories, valeurs)  # barres horizontales (mieux si étiquettes longues)
```

### 🔎 Barres groupées — comparer deux séries

Pour afficher deux séries côte à côte, on décale manuellement les positions des barres avec `numpy` :

```python
import numpy as np
x = np.arange(len(categories))  # positions [0, 1, 2, 3...]
width = 0.35                      # largeur d'une barre

ax.bar(x - width/2, valeurs1, width, label="Série 1")
ax.bar(x + width/2, valeurs2, width, label="Série 2")
ax.set_xticks(x)
ax.set_xticklabels(categories)
```

````python
import matplotlib.pyplot as plt
import numpy as np

produits = ["Pizza", "Burger", "Sushi", "Tacos", "Salade"]
ventes   = [320, 280, 180, 150, 90]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))

# Barres verticales
ax1.bar(produits, ventes, color=["steelblue","coral","seagreen","gold","purple"])
ax1.set_title("Ventes par produit")
ax1.set_xlabel("Produit")
ax1.set_ylabel("Quantité")
ax1.grid(True, axis="y", alpha=0.3)   # grille uniquement sur l'axe Y

# Barres horizontales (mieux si les étiquettes sont longues)
ax2.barh(produits, ventes, color="steelblue")
ax2.set_title("Ventes par produit (horizontal)")
ax2.set_xlabel("Quantité")
ax2.grid(True, axis="x", alpha=0.3)

plt.suptitle("Comparaison des ventes", fontsize=13)
plt.tight_layout()
plt.show()
````

## 📝 Partie 3 – L'histogramme : la distribution

### 🔎 Quand l'utiliser ?
Quand on veut voir **comment se répartissent des valeurs numériques continues** — pas comparer des catégories, mais comprendre la forme de la distribution.

**Questions types :**
- *"Les âges de nos clients sont-ils concentrés ou dispersés ?"*
- *"Y a-t-il beaucoup de notes basses ou la classe est-elle homogène ?"*

```python
plt.hist(donnees, bins=10)    # bins = nombre d'intervalles
plt.hist(donnees, bins=10, edgecolor="white")  # séparation visible entre barres
plt.hist(donnees, bins=10, density=True)       # en % plutôt qu'en nombre
```

### 🔎 Choisir le nombre de bins

```
Trop peu de bins (bins=3) :    Trop de bins (bins=50) :    Bon équilibre (bins=10) :
  ████████                       █ █  █  █ █ █               ██ ████ ████ ██
  ████████ ████                  ██ █  ████ █                ████████████████
  → on perd l'info               → graphique illisible       → on voit la forme
```

Règle pratique : entre **8 et 20 bins** pour la plupart des datasets.

````python
import matplotlib.pyplot as plt
import numpy as np

# Simuler des salaires d'entreprise (distribution réaliste)
np.random.seed(42)
salaires = np.concatenate([
    np.random.normal(2800, 400, 200),   # employés
    np.random.normal(5000, 800, 80),    # cadres
    np.random.normal(9000, 1500, 20)    # direction
])

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Histogramme simple
axes[0].hist(salaires, bins=20, color="steelblue", edgecolor="white")
axes[0].set_title("Distribution des salaires")
axes[0].set_xlabel("Salaire (€)")
axes[0].set_ylabel("Nombre d'employés")
axes[0].axvline(np.mean(salaires),  color="red",    linestyle="--", label=f"Moyenne : {np.mean(salaires):.0f}€")
axes[0].axvline(np.median(salaires),color="orange", linestyle="-.", label=f"Médiane : {np.median(salaires):.0f}€")
axes[0].legend()

# Impact du nombre de bins
axes[1].hist(salaires, bins=5,  color="coral",     edgecolor="white", alpha=0.7, label="bins=5")
axes[1].hist(salaires, bins=40, color="steelblue", edgecolor="white", alpha=0.5, label="bins=40")
axes[1].set_title("Impact du nombre de bins")
axes[1].set_xlabel("Salaire (€)")
axes[1].legend()

plt.suptitle("Histogrammes : distribution des salaires", fontsize=13)
plt.tight_layout()
plt.show()

print(f"Moyenne : {np.mean(salaires):.0f} € — Médiane : {np.median(salaires):.0f} €")
print("→ La moyenne est tirée vers le haut par les salaires élevés de la direction")
````

## 📝 Partie 4 – Le nuage de points : les relations

### 🔎 Quand l'utiliser ?
Quand on veut voir s'il existe une **relation entre deux variables numériques** continues.

**Questions types :**
- *"Les employés plus anciens ont-ils un salaire plus élevé ?"*
- *"Y a-t-il un lien entre budget pub et chiffre d'affaires ?"*

```python
plt.scatter(x, y)                          # nuage simple
plt.scatter(x, y, c=couleurs, cmap="viridis")  # couleur par catégorie
plt.scatter(x, y, s=tailles)               # taille des points proportionnelle
```

### 🔎 Lien avec la corrélation

```
Points en ligne montante  → corrélation positive  (r proche de +1)
Points en ligne descendante → corrélation négative (r proche de -1)
Points éparpillés           → pas de relation     (r proche de 0)
```

On peut calculer le coefficient de corrélation avec `np.corrcoef(x, y)[0,1]`.

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
# Ancienneté (années) et salaire
anciennete = np.random.randint(1, 20, 80)
salaire    = anciennete * 180 + np.random.normal(2500, 400, 80)

fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(anciennete, salaire, color="steelblue", alpha=0.6, edgecolors="white", s=60)

ax.set_title("Relation ancienneté / salaire")
ax.set_xlabel("Ancienneté (années)")
ax.set_ylabel("Salaire (€)")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

r = np.corrcoef(anciennete, salaire)[0, 1]
print(f"Corrélation ancienneté ↔ salaire : {r:.3f}")
print("→ Corrélation forte positive : plus on est ancien, plus on est payé")
````

## 📝 Partie 5 – Le camembert : les proportions

### 🔎 Quand l'utiliser ?
Quand on veut montrer la **répartition en pourcentage** de quelques catégories.

**Question type :** *"Quelle part du CA représente chaque région ?"*

> ⚠️ **Limites :** difficile à lire avec plus de 5-6 catégories. Si les parts sont proches en taille, préférez les barres.

```python
plt.pie(
    valeurs,
    labels    = categories,
    autopct   = "%1.1f%%",   # affiche les pourcentages
    startangle= 90,           # commence à 12h
    explode   = (0.05, 0, 0, 0)  # "sort" la première part
)
```

````python
import matplotlib.pyplot as plt

regions = ["Nord", "Sud", "Est", "Ouest", "Paris"]
ca      = [25, 18, 12, 15, 30]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# Camembert simple
ax1.pie(
    ca,
    labels     = regions,
    autopct    = "%1.1f%%",
    startangle = 90,
    colors     = ["steelblue", "coral", "seagreen", "gold", "purple"]
)
ax1.set_title("Répartition du CA par région")

# Même données en barres — plus lisible pour comparer
ax2.bar(regions, ca, color=["steelblue","coral","seagreen","gold","purple"])
ax2.set_title("Même données en barres")
ax2.set_ylabel("CA (M€)")
ax2.grid(True, axis="y", alpha=0.3)

plt.suptitle("Camembert vs Barres — quand choisir ?", fontsize=13)
plt.tight_layout()
plt.show()
print("→ Camembert : bien pour les proportions / Barres : mieux pour comparer les valeurs exactes")
````

## ✅ Récapitulatif

| Type | Fonction | Quand l'utiliser |
|------|----------|------------------|
| **Courbe** | `plt.plot()` | Évolution dans le temps, tendance |
| **Barres verticales** | `plt.bar()` | Comparer des catégories |
| **Barres horizontales** | `plt.barh()` | Idem, si étiquettes longues |
| **Barres groupées** | `ax.bar(x ± width/2, ...)` | Comparer 2 séries par catégorie |
| **Histogramme** | `plt.hist()` | Distribution d'une variable numérique |
| **Nuage de points** | `plt.scatter()` | Relation entre 2 variables numériques |
| **Camembert** | `plt.pie()` | Répartition en % (max 5-6 catégories) |

**Astuces :**
- `ax.grid(True, axis="y", alpha=0.3)` → grille légère uniquement sur l'axe Y
- `ax.axvline(valeur)` → ligne verticale (ex: moyenne sur histogramme)
- `fig.suptitle()` → titre général au-dessus de tous les panneaux
- `alpha=0.6` sur `scatter` → points semi-transparents (lisible si superposition)
