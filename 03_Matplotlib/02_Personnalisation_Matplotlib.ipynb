## 📝 Partie 1 – Styles globaux

### 🔎 Changer l'apparence générale avec `plt.style.use()`

Matplotlib propose des **styles prédéfinis** qui changent d'un coup la police, les couleurs, le fond, la grille — comme changer de thème dans PowerPoint.

```python
plt.style.use("seaborn-v0_8-whitegrid")  # fond blanc avec grille — très lisible
plt.style.use("ggplot")                   # style R ggplot — couleurs douces
plt.style.use("dark_background")          # fond noir — pour présentations sombres
plt.style.use("fivethirtyeight")          # style journalistique — titres grands
plt.style.use("default")                  # style par défaut Matplotlib

# Voir tous les styles disponibles :
print(plt.style.available)
```

> 💡 **Bonne pratique en entreprise :** choisissez un style au début du notebook et gardez-le pour toute la session. Cela garantit une cohérence visuelle entre tous les graphiques du rapport.

````python
import matplotlib.pyplot as plt
import numpy as np

mois = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun"]
ca   = [42, 38, 55, 60, 72, 85]

styles = ["default", "ggplot", "seaborn-v0_8-whitegrid", "fivethirtyeight"]

fig, axes = plt.subplots(2, 2, figsize=(13, 8))

for ax, style in zip(axes.flat, styles):
    with plt.style.context(style):          # applique le style seulement pour ce bloc
        ax.plot(mois, ca, marker="o", linewidth=2)
        ax.set_title(f"Style : '{style}'")
        ax.set_xlabel("Mois")
        ax.set_ylabel("CA (k€)")

fig.suptitle("Comparaison des styles Matplotlib", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()
````

## 📝 Partie 2 – Contrôler les axes

### 🔎 Limites, graduations et grille

Par défaut, Matplotlib choisit automatiquement les échelles. Mais on a souvent besoin de les contrôler manuellement :

```python
# Limites des axes
ax.set_xlim(0, 10)         # axe X de 0 à 10
ax.set_ylim(0, 100)        # axe Y de 0 à 100

# Graduations personnalisées
ax.set_xticks([0, 2, 4, 6, 8, 10])              # positions des graduations
ax.set_xticklabels(["Jan", "Mar", "Mai", ...])   # étiquettes des graduations

# Rotation des étiquettes (quand elles se chevauchent)
ax.tick_params(axis="x", rotation=45)

# Grille
ax.grid(True)                              # grille simple
ax.grid(True, axis="y", alpha=0.4)        # grille légère sur Y seulement
ax.grid(True, linestyle="--", color="gray", alpha=0.5)  # grille personnalisée

# Lignes de référence horizontales et verticales
ax.axhline(y=50,  color="red",   linestyle="--", label="Objectif")  # ligne horiz.
ax.axvline(x=2.5, color="green", linestyle=":",  label="Seuil")     # ligne vertic.
```

````python
import matplotlib.pyplot as plt
import numpy as np

mois = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun",
        "Jul", "Aoû", "Sep", "Oct", "Nov", "Déc"]
ca   = [42, 38, 55, 60, 72, 85, 90, 78, 65, 70, 88, 95]
objectif = 70   # objectif mensuel en k€

fig, ax = plt.subplots(figsize=(11, 5))

ax.plot(mois, ca, color="steelblue", marker="o", linewidth=2, label="CA réel")

# Ligne d'objectif
ax.axhline(y=objectif, color="red", linestyle="--", linewidth=1.5,
           label=f"Objectif : {objectif} k€")

# Zones colorées : mois sous l'objectif
for i, (m, v) in enumerate(zip(mois, ca)):
    if v < objectif:
        ax.axvspan(i - 0.3, i + 0.3, color="red", alpha=0.1)   # zone rouge légère

# Contrôle des axes
ax.set_ylim(25, 110)                        # on force l'axe Y à commencer à 25
ax.tick_params(axis="x", rotation=45)       # étiquettes X à 45° pour éviter la superposition
ax.grid(True, axis="y", alpha=0.3, linestyle="--")

ax.set_title("CA mensuel vs objectif", fontsize=13)
ax.set_xlabel("Mois")
ax.set_ylabel("CA (k€)")
ax.legend()
plt.tight_layout()
plt.show()
````

## 📝 Partie 3 – Annotations : mettre en valeur l'information

### 🔎 `ax.annotate()` — texte + flèche sur un point précis

Les annotations permettent d'attirer l'attention sur un point particulier — le pic de ventes, une anomalie, une date clé.

```python
ax.annotate(
    "Mon texte",              # texte à afficher
    xy     = (x, y),          # point vers lequel pointe la flèche
    xytext = (x+0.5, y+5),   # position du texte
    arrowprops = dict(
        arrowstyle = "->",    # style de la flèche
        color      = "red"
    ),
    fontsize = 10,
    color    = "red"
)
```

### 🔎 `ax.text()` — texte simple sans flèche

```python
ax.text(
    x, y,               # position du texte
    "Mon texte",
    fontsize  = 10,
    ha        = "center",   # alignement horizontal : center, left, right
    va        = "bottom",   # alignement vertical : bottom, top, center
    color     = "gray",
    fontweight= "bold"
)
```

### 🔎 Afficher les valeurs sur des barres

Très courant en entreprise — afficher la valeur exacte au-dessus de chaque barre :

```python
for i, (cat, val) in enumerate(zip(categories, valeurs)):
    ax.text(i, val + 2, str(val), ha="center", fontsize=9)
```

````python
import matplotlib.pyplot as plt
import numpy as np

mois = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun"]
ca   = [42, 38, 55, 60, 72, 85]

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(mois, ca, color="steelblue", marker="o", linewidth=2)

# Annoter le pic (dernier mois)
ax.annotate(
    f"Record : {max(ca)} k€",
    xy      = (mois.index(mois[-1]), max(ca)),  # position du point
    xytext  = (3.5, 88),                         # position du texte
    arrowprops = dict(arrowstyle="->", color="red", lw=1.5),
    fontsize = 11,
    color    = "red",
    fontweight = "bold"
)

# Annoter le creux
ax.annotate(
    "Mois creux",
    xy      = (1, 38),
    xytext  = (2, 32),
    arrowprops = dict(arrowstyle="->", color="gray"),
    fontsize = 10,
    color    = "gray"
)

ax.set_title("CA mensuel avec annotations")
ax.set_xlabel("Mois"); ax.set_ylabel("CA (k€)")
ax.set_ylim(25, 100)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt

produits = ["Pizza", "Burger", "Sushi", "Tacos", "Salade"]
ventes   = [320, 280, 180, 150, 90]

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(produits, ventes,
              color=["steelblue","coral","seagreen","gold","purple"])

# Afficher la valeur au-dessus de chaque barre
for bar, val in zip(bars, ventes):
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # centre horizontal de la barre
        bar.get_height() + 5,               # juste au-dessus de la barre
        str(val),
        ha       = "center",
        va       = "bottom",
        fontsize = 11,
        fontweight = "bold"
    )

ax.set_title("Ventes par produit — valeurs affichées")
ax.set_ylabel("Quantité")
ax.set_ylim(0, 380)   # laisser de la place pour les étiquettes
ax.grid(True, axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
````

## 📝 Partie 4 – Couleurs avancées

### 🔎 Spécifier les couleurs

Matplotlib accepte plusieurs façons de définir une couleur :

```python
color = "steelblue"      # nom CSS
color = "#1f77b4"        # code hexadécimal
color = (0.12, 0.47, 0.71)  # tuple RGB normalisé (valeurs entre 0 et 1)
color = "C0"             # couleur du cycle automatique Matplotlib (C0, C1, C2...)
```

### 🔎 Colorer des barres selon leurs valeurs

Très utile pour mettre en évidence les valeurs au-dessus/dessous d'un seuil :

```python
couleurs = ["seagreen" if v >= seuil else "crimson" for v in valeurs]
ax.bar(categories, valeurs, color=couleurs)
```

### 🔎 Transparence avec `alpha`

```python
ax.bar(x, y, alpha=0.7)          # 70% opaque
ax.scatter(x, y, alpha=0.5)      # utile si les points se superposent
ax.fill_between(x, y1, y2, alpha=0.2)  # zone colorée entre deux courbes
```

### 🔎 `fill_between` — zone entre deux courbes

```python
ax.fill_between(
    x,              # valeurs de l'axe X
    y1,             # courbe du bas
    y2,             # courbe du haut
    alpha = 0.2,    # transparence
    color = "steelblue",
    label = "Intervalle"
)
```

````python
import matplotlib.pyplot as plt
import numpy as np

mois    = ["Jan","Fév","Mar","Avr","Mai","Jun","Jul","Aoû","Sep","Oct","Nov","Déc"]
ca_reel = [42, 38, 55, 60, 72, 85, 90, 78, 65, 70, 88, 95]
ca_prev = [50, 50, 60, 65, 70, 80, 85, 80, 70, 75, 85, 100]  # prévisionnel
x       = range(len(mois))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barres colorées selon performance vs objectif
objectif = 70
couleurs = ["seagreen" if v >= objectif else "crimson" for v in ca_reel]
axes[0].bar(mois, ca_reel, color=couleurs, alpha=0.8)
axes[0].axhline(objectif, color="navy", linestyle="--", label=f"Objectif {objectif}k€")
axes[0].set_title("CA : vert=objectif atteint, rouge=manqué")
axes[0].tick_params(axis="x", rotation=45)
axes[0].legend()
axes[0].grid(True, axis="y", alpha=0.3)

# Courbes avec zone entre réel et prévisionnel
axes[1].plot(mois, ca_reel, color="steelblue", marker="o", label="Réel", linewidth=2)
axes[1].plot(mois, ca_prev, color="orange",    marker="s", label="Prévisionnel",
             linewidth=2, linestyle="--")
axes[1].fill_between(x, ca_reel, ca_prev,
                     where=[r < p for r, p in zip(ca_reel, ca_prev)],
                     color="red",   alpha=0.15, label="Sous prévision")
axes[1].fill_between(x, ca_reel, ca_prev,
                     where=[r >= p for r, p in zip(ca_reel, ca_prev)],
                     color="green", alpha=0.15, label="Sur prévision")
axes[1].set_title("Réel vs Prévisionnel")
axes[1].tick_params(axis="x", rotation=45)
axes[1].legend(fontsize=8)
axes[1].grid(True, alpha=0.3)

plt.suptitle("Personnalisation avancée des couleurs", fontsize=13)
plt.tight_layout()
plt.show()
````

### 🔎 La légende

```python

    loc       = "upper left",  # position : upper/lower + left/right/center
    fontsize  = 10,
    framealpha= 0.9,           # transparence du cadre
    shadow    = True,
    ncol      = 2              # légende sur 2 colonnes
)
```

````python
import matplotlib.pyplot as plt

mois   = ["Jan", "Fév", "Mar", "Avr", "Mai", "Jun"]
ventes = [120, 100, 145, 160, 190, 230]
couts  = [80,   75, 100, 110, 130, 155]

fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(mois, ventes, color="steelblue", marker="o", linewidth=2, label="Ventes")
ax.plot(mois, couts,  color="coral",     marker="s", linewidth=2, label="Coûts", linestyle="--")

# Titre bien formaté
ax.set_title(
    "Ventes vs Coûts — Semestre 1",
    fontsize=14, fontweight="bold", color="#222222", pad=15
)

# Labels d'axes
ax.set_xlabel("Mois", fontsize=11, labelpad=8)
ax.set_ylabel("Montant (k€)", fontsize=11)

# Légende positionnée
ax.legend(loc="upper left", fontsize=10, framealpha=0.9)

# Taille des graduations
ax.tick_params(axis="both", labelsize=10)

ax.grid(True, alpha=0.3, linestyle=":")
plt.tight_layout()
plt.show()
````

    xytext=(idx_max - 2, ca_france[idx_max] + 30),
    arrowprops=dict(arrowstyle="->", color="red"),
    color="red", fontweight="bold")

ax.set_title("CA France vs Export — 2023/2024",
             fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("Trimestre")

ax.legend(loc="upper left", fontsize=10)
ax.tick_params(axis="x", rotation=45)
ax.grid(True, alpha=0.3, linestyle=":")

plt.tight_layout()
plt.savefig("rapport_ca.png", dpi=150, bbox_inches="tight")

## ✅ Récapitulatif

| Concept | Code | Ce qu'il faut retenir |
|---------|------|-----------------------|
| **Style global** | `plt.style.use("ggplot")` | Change l'apparence de tous les graphiques du notebook |

| **Ligne verticale** | `ax.axvline(x=val)` | Date clé, rupture |
| **Grille** | `ax.grid(True, axis="y", alpha=0.3, linestyle=":")` | Légère et discrète |
| **Annotation + flèche** | `ax.annotate("texte", xy=..., xytext=..., arrowprops=...)` | Mettre en valeur un point |

| **Zone entre courbes** | `ax.fill_between(x, y1, y2, alpha=0.2)` | Visualiser un écart |
| **Couleurs conditionnelles** | `["vert" if v>seuil else "rouge" for v in vals]` | Colorer selon valeur |
