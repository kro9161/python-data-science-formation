## 📝 Partie 1 – Pourquoi Matplotlib ?

### 🔎 La place de Matplotlib dans l'écosystème

On a vu dans les notebooks Pandas que `df.plot()` permet de créer des graphiques rapidement. Mais `df.plot()` est limité — il s'appuie en coulisse sur Matplotlib.

**Matplotlib, c'est la couche inférieure** : le moteur de dessin sur lequel reposent Pandas, Seaborn, et la plupart des bibliothèques de visualisation Python.

```
Niveau 3 — Seaborn          → graphiques statistiques beaux par défaut
Niveau 2 — Pandas .plot()   → visualisation rapide depuis un DataFrame
Niveau 1 — Matplotlib       → contrôle total sur chaque détail du graphique
```

**Quand utiliser Matplotlib directement ?**
- Quand on veut contrôler **précisément** chaque élément visuel
- Quand on crée des **graphiques multi-panneaux** complexes
- Quand on veut **annoter** un graphique (flèches, textes, zones colorées)
- Quand `df.plot()` ne suffit plus

### 🔎 Importer Matplotlib

Par convention, on importe toujours `pyplot` avec l'alias `plt` :

```python
import matplotlib.pyplot as plt
```

`pyplot` est le sous-module de Matplotlib qui contient toutes les fonctions de dessin. C'est la seule ligne d'import dont vous aurez besoin dans 95% des cas.

## 📝 Partie 2 – Premier graphique

### 🔎 La logique de base

Tracer un graphique avec Matplotlib se fait en 3 étapes :

```
1. Préparer les données    → deux listes (x et y)
2. Tracer                  → plt.plot(x, y)
3. Afficher                → plt.show()
```

C'est tout. Matplotlib s'occupe du reste automatiquement (échelles, graduations, taille…).

````python
import matplotlib.pyplot as plt

# Données : températures sur 5 jours
jours        = ["Lun", "Mar", "Mer", "Jeu", "Ven"]
temperatures = [15, 17, 16, 14, 18]

# Tracer la courbe
plt.plot(jours, temperatures)

# Afficher
plt.show()
````

### 🔎 Ce que fait chaque ligne

```python
plt.plot(jours, temperatures)
#         ↑           ↑
#    axe X (horizontal)  axe Y (vertical)
#    les étiquettes       les valeurs à tracer

plt.show()
# → affiche le graphique dans Jupyter
# → sans cette ligne, rien ne s'affiche (ou texte parasite)
```

> ⚠️ **`plt.show()` est important** : il vide le "pinceau" de Matplotlib et affiche le graphique. Sans lui dans certains environnements, le graphique ne s'affiche pas proprement.

## 📝 Partie 3 – Personnaliser un graphique

### 🔎 Les éléments essentiels

Un graphique sans titre ni labels d'axes est difficile à lire pour quelqu'un qui ne connaît pas les données. Les éléments minimaux à toujours ajouter :

```python
plt.title("Mon titre")       # titre du graphique
plt.xlabel("Axe horizontal") # label axe X
plt.ylabel("Axe vertical")   # label axe Y
plt.legend()                  # légende (si plusieurs courbes)
```

### 🔎 Personnaliser l'apparence de la courbe

```python
plt.plot(x, y,
    color     = "steelblue",  # couleur : nom ou code hex (#1f77b4)
    linestyle = "--",          # style : "-" plein, "--" tirets, ":" pointillés, "-." mixte
    linewidth = 2,             # épaisseur de la ligne
    marker    = "o",           # marqueur sur chaque point : "o" rond, "s" carré, "^" triangle
    markersize= 8,             # taille des marqueurs
    label     = "Température" # texte pour la légende
)
```

**Couleurs courantes :**

| Nom | Couleur | Nom | Couleur |
|-----|---------|-----|--------|
| `"steelblue"` | Bleu acier | `"coral"` | Corail |
| `"orange"` | Orange | `"seagreen"` | Vert |
| `"crimson"` | Rouge | `"purple"` | Violet |
| `"gray"` | Gris | `"gold"` | Or |

````python
import matplotlib.pyplot as plt

jours        = ["Lun", "Mar", "Mer", "Jeu", "Ven"]
temperatures = [15, 17, 16, 14, 18]

plt.plot(
    jours, temperatures,
    color     = "steelblue",
    linestyle = "-",
    linewidth = 2,
    marker    = "o",
    markersize = 8,
    label     = "Température (°C)"
)

plt.title("Évolution de la température sur la semaine")
plt.xlabel("Jour")
plt.ylabel("Température (°C)")
plt.legend()
plt.tight_layout()   # évite que les étiquettes soient coupées
plt.show()
````

## 📝 Partie 4 – `plt.xxx` vs `fig, ax` : deux styles d'écriture

### 🔎 C'est LA notion fondamentale de Matplotlib

Vous verrez deux façons d'écrire du code Matplotlib dans les tutoriels et la documentation. Elles produisent **le même résultat** mais fonctionnent différemment.

**Style 1 — `plt.xxx` (style fonctionnel) :**
```python
plt.plot(x, y)
plt.title("Mon titre")
plt.show()
```
→ Simple, pratique pour un graphique rapide. Matplotlib gère tout en coulisse.

**Style 2 — `fig, ax` (style objet) :**
```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Mon titre")
plt.show()
```
→ Plus verbeux, mais **indispensable** pour les graphiques multi-panneaux et le contrôle fin.

### 🔎 L'analogie : la toile et le pinceau

```
fig = la TOILE (le fond, la feuille entière)
 ax = le PANNEAU de dessin sur la toile
      (on peut en avoir plusieurs sur une même toile)

fig, ax = plt.subplots()   → crée 1 toile avec 1 panneau
fig, axes = plt.subplots(2, 3)  → crée 1 toile avec 6 panneaux (grille 2×3)
```

### 🔎 Équivalences `plt` ↔ `ax`

| Style `plt` | Style `ax` | Ce que ça fait |
|-------------|-----------|----------------|
| `plt.plot()` | `ax.plot()` | Trace une courbe |
| `plt.title("titre")` | `ax.set_title("titre")` | Titre du graphique |
| `plt.xlabel("x")` | `ax.set_xlabel("x")` | Label axe X |
| `plt.ylabel("y")` | `ax.set_ylabel("y")` | Label axe Y |
| `plt.legend()` | `ax.legend()` | Légende |
| `plt.grid()` | `ax.grid()` | Grille |

> 💡 **Recommandation :** utilisez `plt.xxx` pour les graphiques simples (1 panneau). Utilisez `fig, ax` dès que vous voulez plusieurs panneaux ou un contrôle précis.

````python
import matplotlib.pyplot as plt

jours        = ["Lun", "Mar", "Mer", "Jeu", "Ven"]
temperatures = [15, 17, 16, 14, 18]
humidite     = [70, 60, 75, 80, 65]

# ===== Style plt (simple) =====
plt.figure(figsize=(8, 4))
plt.plot(jours, temperatures, color="steelblue", marker="o", label="Temp. (°C)")
plt.title("Style plt — simple")
plt.xlabel("Jour")
plt.ylabel("Valeur")
plt.legend()
plt.tight_layout()
plt.show()

# ===== Style fig, ax (objet) =====
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(jours, temperatures, color="steelblue", marker="o", label="Temp. (°C)")
ax.set_title("Style fig, ax — objet")
ax.set_xlabel("Jour")
ax.set_ylabel("Valeur")
ax.legend()
plt.tight_layout()
plt.show()

print("→ Les deux graphiques sont identiques — deux façons d'écrire la même chose")
````

````python
import matplotlib.pyplot as plt

jours        = ["Lun", "Mar", "Mer", "Jeu", "Ven"]
temperatures = [15, 17, 16, 14, 18]
humidite     = [70, 60, 75, 80, 65]

# Style fig, ax — avantage : 2 panneaux côte à côte sur la même toile
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Panneau gauche : température
ax1.plot(jours, temperatures, color="steelblue", marker="o")
ax1.set_title("Température")
ax1.set_xlabel("Jour")
ax1.set_ylabel("°C")

# Panneau droit : humidité
ax2.plot(jours, humidite, color="coral", marker="s", linestyle="--")
ax2.set_title("Humidité")
ax2.set_xlabel("Jour")
ax2.set_ylabel("%")

fig.suptitle("Météo de la semaine", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
````

## 📝 Partie 5 – Sauvegarder un graphique

### 🔎 `plt.savefig()`

```python
plt.savefig("nom_fichier.png")          # PNG par défaut
plt.savefig("rapport.pdf")              # PDF vectoriel (idéal pour impression)
plt.savefig("graphique.svg")            # SVG vectoriel (idéal pour le web)
plt.savefig("qualite.png", dpi=300)     # haute résolution (300 dpi = qualité impression)
plt.savefig("propre.png", bbox_inches="tight")  # pas de coupure des étiquettes
```

> ⚠️ **Important :** `plt.savefig()` doit être appelé **avant** `plt.show()`. Si vous appelez `plt.show()` d'abord, Matplotlib vide le graphique et `savefig()` sauvegarde une image vide.

```python
# ✅ BON ordre
plt.plot(...)
plt.savefig("graphique.png")   # sauvegarder AVANT
plt.show()                      # afficher APRÈS

# ❌ MAUVAIS ordre
plt.plot(...)
plt.show()                      # vide le graphique
plt.savefig("graphique.png")   # sauvegarde une image vide !
```

````python
import matplotlib.pyplot as plt

jours        = ["Lun", "Mar", "Mer", "Jeu", "Ven"]
temperatures = [15, 17, 16, 14, 18]

plt.figure(figsize=(8, 4))
plt.plot(jours, temperatures, color="steelblue", marker="o", linewidth=2)
plt.title("Températures de la semaine")
plt.xlabel("Jour")
plt.ylabel("°C")
plt.tight_layout()

# Sauvegarder AVANT plt.show()
plt.savefig("temperatures.png", dpi=150, bbox_inches="tight")
print("✅ Sauvegardé : temperatures.png")

plt.show()
````

## ⚠️ Erreurs fréquentes

| Problème | Cause | Solution |
|----------|-------|----------|
| Rien ne s'affiche | Oubli de `plt.show()` | Toujours finir par `plt.show()` |
| Texte `<Figure size...>` affiché | Pas de `plt.show()` | Ajouter `plt.show()` |
| Étiquettes coupées | Pas de `tight_layout` | Ajouter `plt.tight_layout()` avant `show` |
| Fichier sauvegardé vide | `savefig` après `show` | Mettre `savefig` **avant** `show` |
| Deux graphiques superposés | Oubli de `plt.figure()` | Commencer par `plt.figure()` ou `plt.subplots()` |
| `ax.title()` ne marche pas | Mauvaise méthode | Utiliser `ax.set_title()` (avec `set_`) |

## ✅ Récapitulatif

| Concept | Code | Ce qu'il faut retenir |
|---------|------|-----------------------|
| **Importer** | `import matplotlib.pyplot as plt` | Toujours cet alias |
| **Tracer** | `plt.plot(x, y)` | x = axe horizontal, y = axe vertical |
| **Afficher** | `plt.show()` | Toujours en dernier |
| **Titre** | `plt.title("titre")` | |
| **Labels axes** | `plt.xlabel()` / `plt.ylabel()` | |
| **Couleur** | `color="steelblue"` | Paramètre de `plt.plot()` |
| **Style ligne** | `linestyle="--"` | `-` `-\-` `:` `-.` |
| **Marqueur** | `marker="o"` | `o` `s` `^` `x` `+` |
| **Taille figure** | `plt.figure(figsize=(8,4))` | Avant de tracer |
| **Style objet** | `fig, ax = plt.subplots()` | Pour multi-panneaux |
| **Éviter coupures** | `plt.tight_layout()` | Avant `show` ou `savefig` |
| **Sauvegarder** | `plt.savefig("f.png", dpi=150)` | Avant `plt.show()` |

**La règle d'or :** `plt.xxx` pour 1 graphique simple, `fig, ax` dès qu'on veut plusieurs panneaux ou un contrôle précis.
