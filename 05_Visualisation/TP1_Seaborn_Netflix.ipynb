## 🧠 1. Comprendre Seaborn et Matplotlib

Seaborn repose sur Matplotlib.  
Tu peux voir Matplotlib comme **le moteur**, et Seaborn comme **le tableau de bord stylé**.

| Élément | Rôle |
|----------|------|
| **Matplotlib** (`plt`) | gère le dessin bas niveau, les axes, la figure, la sauvegarde |
| **Seaborn** (`sns`) | ajoute un style moderne, une logique statistique et des graphes plus intuitifs |

✅ On importe les deux :
```python
import seaborn as sns
import matplotlib.pyplot as plt
```


````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 🌈 Appliquer un thème visuel agréable
sns.set_theme(style="whitegrid")
````

🧠 **Mots-clés :**
- `sns` → alias pour Seaborn.
- `plt` → alias pour Matplotlib (titres, axes, sauvegarde).
- `set_theme()` → applique un style global (fond clair + grille).

````
df = pd.read_csv("netflix_titles.csv")
df.head()
````

🧠 **Explications :**
- `pd.read_csv()` → charge un fichier CSV en DataFrame.
- `head()` → affiche les 5 premières lignes pour vérifier le chargement.

````
sns.countplot(x="type", data=df, palette="Set2")
plt.title("Répartition des types de contenu sur Netflix")
plt.xlabel("Type de contenu")
plt.ylabel("Nombre de titres")
plt.show()
````

🧠 **Mots-clés :**
- `countplot()` → compte automatiquement les occurrences d’une catégorie.
- `x` → variable en abscisse.
- `palette` → palette de couleurs Seaborn.
- `plt.show()` → affiche le graphique.

````
df_films = df[df["type"] == "Movie"].copy()
df_films["duration_min"] = df_films["duration"].str.replace(" min", "").astype(float)

sns.histplot(df_films["duration_min"], bins=10, kde=True, color="salmon")
plt.title("Distribution des durées des films Netflix")
plt.xlabel("Durée (minutes)")
plt.ylabel("Fréquence")
plt.show()
````

🧠 **Mots-clés :**
- `histplot()` → histogramme (répartition).
- `bins=10` → nombre de classes.
- `kde=True` → courbe de densité.
- `str.replace()` / `astype(float)` → convertit les durées en nombres.

````
top_pays = df_films["country"].value_counts().head(5).index

sns.boxplot(x="country", y="duration_min",
            data=df_films[df_films["country"].isin(top_pays)],
            palette="pastel")
plt.title("Durée des films selon le pays (Top 5)")
plt.xlabel("Pays")
plt.ylabel("Durée (min)")
plt.xticks(rotation=45)
plt.show()
````

🧠 **Mots-clés :**
- `boxplot()` → montre la distribution (médiane, quartiles, extrêmes).
- `xticks(rotation=45)` → fait pivoter les étiquettes pour plus de lisibilité.

````
sns.scatterplot(x="release_year", y="duration_min", data=df_films, color="purple")
plt.title("Relation entre année de sortie et durée des films")
plt.xlabel("Année de sortie")
plt.ylabel("Durée (minutes)")
plt.show()
````

🧠 **Mots-clés :**
- `scatterplot()` → nuage de points.
- Chaque point = un film : année ↔ durée.

````
corr = df_films[["release_year", "duration_min"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Corrélation entre année et durée des films Netflix")
plt.show()
````

🧠 **Mots-clés :**
- `corr()` → calcule la corrélation.
- `heatmap()` → carte de chaleur.
- `annot=True` → affiche les valeurs.
- `fmt=".2f"` → format à deux décimales.

````
plt.figure(figsize=(8,5))
sns.boxplot(x="country", y="duration_min",
            data=df_films[df_films["country"].isin(top_pays)],
            palette="Set3")
plt.title("Durée moyenne des films par pays (Netflix)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("netflix_boxplot.png", dpi=300)
plt.show()
````
