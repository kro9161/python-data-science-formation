fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1️⃣ Évolution par jour
sns.lineplot(x="Jour", y="Pizzas_vendues", data=df, ax=axes[0], color="tomato")
axes[0].set_title("Évolution des ventes")

# 2️⃣ Impact météo
sns.barplot(x="Pluie", y="Pizzas_vendues", data=df, ax=axes[1], palette="Set2")
axes[1].set_title("Impact météo")

# 3️⃣ Corrélations
sns.heatmap(df.select_dtypes("number").corr(), annot=True, cmap="coolwarm", ax=axes[2])
axes[2].set_title("Corrélations")

plt.suptitle("🍕 Dashboard final des ventes de pizzas", fontsize=16)
plt.tight_layout()
plt.show()
