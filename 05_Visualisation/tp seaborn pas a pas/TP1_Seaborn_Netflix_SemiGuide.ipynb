## 🧠 1) Comprendre Seaborn et Matplotlib

**Seaborn** repose sur **Matplotlib**.  
Pense-le ainsi : Matplotlib = moteur, Seaborn = interface moderne.

|----------|------|
| **Matplotlib (`plt`)** | Gère le dessin bas niveau, les axes, la figure, la sauvegarde |
| **Seaborn (`sns`)** | Apporte un style moderne et des fonctions statistiques plus simples |

## 🧠 Message clé

Avec **Seaborn**, tu peux visualiser rapidement des relations statistiques :
- `countplot` pour les catégories  
- `histplot` pour les distributions  
- `boxplot` pour comparer des groupes  
- `scatterplot` pour des relations numériques
