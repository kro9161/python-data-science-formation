## 🧩 5) Vue d’ensemble : `pairplot()` 🔍

````python
# Ton code ici
````


<details><summary>✅ Correction</summary>

```python
g = sns.pairplot(
    df[["Pizzas_vendues","Prix_moyen","Pluie"]],
    hue="Pluie",

    palette="coolwarm"

plt.show()

🧠 **Mots-clés** : `pairplot` dessine toutes les combinaisons ; `diag_kind="kde"` ajoute une densité sur la diagonale.

g.map_dataframe(sns.barplot, x="Jour", y="Pizzas_vendues", ci=None, palette="muted")
g.set_titles("Pluie = {col_name}")
g.set_axis_labels("Jour", "Pizzas vendues")
plt.suptitle("Ventes journalières selon la météo", y=1.05)
plt.show()
```
