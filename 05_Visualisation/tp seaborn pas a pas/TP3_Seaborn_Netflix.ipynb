## 🧩 2. Préparer les données  

💪 *À toi de jouer :*  
- Garde les colonnes `['type', 'release_year', 'duration', 'country', 'rating']`.  
- Supprime les lignes vides (`dropna`).  
- Crée une colonne numérique `duree_minutes` à partir de `'duration'`.  


````
# Ton code ici
````


<details><summary>✅ Correction</summary>

```python
df = df[['type', 'release_year', 'duration', 'country', 'rating']].dropna()
df['duree_minutes'] = df['duration'].str.replace(' min', '', regex=False).astype(float)
df.head()
```
</details>

## 🧩 4. Durée moyenne des films par année (Boxplot) 🎞️  

💪 *À toi de jouer :*

- Crée un `boxplot` avec :  
  - X = `release_year`  
  - Y = `duree_minutes`  
  - `palette='coolwarm'`  
- Limite aux années récentes (> 2010).  


````
# Ton code ici
````


<details><summary>✅ Correction</summary>

```python
films = df[df['type'] == 'Movie']
recent = films[films['release_year'] > 2010]

plt.figure(figsize=(10,5))
sns.boxplot(x='release_year', y='duree_minutes', data=recent, palette='coolwarm')
plt.title('Durée des films Netflix récents')
plt.xlabel('Année de sortie')
plt.ylabel('Durée (minutes)')
plt.xticks(rotation=45)
plt.show()
```

## 🧩 6. Corrélation entre variables numériques 🔢  

💪 *À toi de jouer :*  
- Calcule la corrélation entre `release_year` et `duree_minutes`.  
- Affiche la matrice avec `sns.heatmap()`.  


````
# Ton code ici
````


<details><summary>✅ Correction</summary>

```python
corr = df[['release_year', 'duree_minutes']].corr()

plt.figure(figsize=(4,3))
sns.heatmap(corr, annot=True, cmap='viridis', fmt='.2f')
plt.title('Corrélation entre année et durée')
plt.show()
```
</details>



## 🧩 7. Top 10 des pays producteurs 🌍

- Trace un `barplot` horizontal.  


````
# Ton code ici
````


<details><summary>✅ Correction</summary>

```python
top_pays = (
    df['country']
    .value_counts()
    .head(10)
    .reset_index()
    .rename(columns={'index': 'Pays', 'country': 'Nombre'})
)

plt.figure(figsize=(8,5))
sns.barplot(y='Pays', x='Nombre', data=top_pays, palette='crest')
plt.title('Top 10 des pays producteurs de contenus Netflix')
plt.xlabel('Nombre de titres')
plt.ylabel('')
plt.show()
```
</details>
