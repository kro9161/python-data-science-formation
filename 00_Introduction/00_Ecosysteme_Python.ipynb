## 📝 Partie 1 – Pourquoi Python pour la Data Science ?

Python n'a pas été conçu pour la data science. C'est un langage **généraliste**, créé en 1991 par Guido van Rossum. Pourtant, il est aujourd'hui le **langage numéro 1** dans ce domaine.

### Pourquoi ?

| Raison | Explication |
|--------|-------------|
| **Lisibilité** | Syntaxe proche du pseudo-code, facile à lire et maintenir |
| **Écosystème riche** | Des centaines de librairies spécialisées (NumPy, Pandas, Matplotlib…) |
| **Communauté massive** | Millions de développeurs, documentation abondante, Stack Overflow |
| **Polyvalence** | Utilisable du script de 10 lignes au système de ML en production |
| **Interopérabilité** | S'interface avec C, Fortran, R, Julia, SQL, Java… |
| **Gratuité** | Open source, aucune licence à payer |

### Python vs ses concurrents

| Langage | Points forts | Limites |
|---------|-------------|--------|
| **Python** | Lisible, riche, polyvalent | Plus lent que C/Java en natif |
| **R** | Statistiques poussées | Moins général, moins utilisé en prod |
| **Julia** | Très rapide | Écosystème encore jeune |
| **Scala** | Big Data (Spark) | Courbe d'apprentissage élevée |

> 💡 **En pratique** : Python gagne car il permet de faire **tout le pipeline** : collecte, nettoyage, analyse, visualisation, ML, déploiement.

## 📝 Partie 3 – Anaconda et la gestion des environnements

### 3.1 – Qu'est-ce qu'Anaconda ?

**Anaconda** est une **distribution Python** spécialisée pour la Data Science. Elle inclut :
- Python
- Plus de **250 librairies scientifiques** préinstallées
- Le gestionnaire de paquets **conda**
- **JupyterLab** et **Jupyter Notebook**
- **Anaconda Navigator** (interface graphique)

**Miniconda** = version légère : Python + conda uniquement, on installe le reste au besoin.

### 3.2 – Pourquoi utiliser un environnement virtuel ?

**Problème concret** : vous avez deux projets :
- Projet A requiert `pandas 1.3` (ancienne API)
- Projet B requiert `pandas 2.1` (nouvelle API)

Si vous installez tout dans le même Python → **conflit de versions** → tout casse.

**Solution** : chaque projet a son propre **environnement virtuel** = une installation Python isolée.

```
  Votre machine
  ├── env_projetA/  →  Python 3.9 + pandas 1.3 + numpy 1.21
  ├── env_projetB/  →  Python 3.11 + pandas 2.1 + numpy 1.26
  └── env_formation/ →  Python 3.12 + toutes les libs du cours
```

### 3.3 – Commandes conda essentielles

```bash
# Créer un environnement nommé "formation" avec Python 3.12
conda create -n formation python=3.12

# Activer l'environnement
conda activate formation

# Installer des librairies dans l'environnement actif
conda install numpy pandas matplotlib seaborn

# Lister les environnements disponibles
conda env list

# Exporter l'environnement (pour le partager ou le reproduire)
conda env export > environment.yml

# Recréer un environnement depuis un fichier .yml
conda env create -f environment.yml

# Désactiver l'environnement actif
conda deactivate

# Supprimer un environnement
conda env remove -n formation
```

> ⚠️ **Bonne pratique** : ne jamais installer de librairies dans l'environnement `base`. Toujours créer un environnement dédié par projet.

### 3.4 – conda vs pip : quelle différence ?

| | **conda** | **pip** |
|-|-----------|---------|
| Gestionnaire de | Paquets Python **et** non-Python (C, Fortran…) | Paquets Python uniquement |
| Résolution de dépendances | Plus robuste pour les librairies scientifiques | Parfois des conflits |
| Source des paquets | Dépôts Anaconda (curatés) | PyPI (tout le monde peut publier) |
| Recommandation | Préférer conda pour NumPy, SciPy, etc. | Utiliser pip si le paquet n'est pas sur conda |

```bash
# Installer avec pip dans un environnement conda (en dernier recours)
pip install nom_du_paquet
```

## 📝 Partie 4 – IPython et Jupyter

### 4.1 – IPython : Python interactif amélioré

**IPython** (Interactive Python) est un interpréteur Python enrichi. Il ajoute :

| Fonctionnalité | Exemple |
|---------------|--------|
| **Autocomplétion** | Tapez `np.` puis `Tab` → liste des méthodes |
| **Aide rapide** | `np.array?` → docstring ; `np.array??` → code source |
| **Commandes magiques** | `%timeit`, `%matplotlib inline`, `%run script.py` |
| **Historique** | Flèche haut/bas, `%history` |
| **Shell intégré** | `!ls`, `!pip install pandas` |

### 4.2 – Jupyter Notebook vs JupyterLab

| | **Jupyter Notebook** | **JupyterLab** |
|-|---------------------|---------------|
| Interface | Simple, un notebook à la fois | IDE complet, onglets multiples |
| Explorateur de fichiers | Limité | Intégré à gauche |
| Terminal intégré | Non | Oui |
| Recommandé | Pour débuter | Pour travailler au quotidien |

```bash
# Lancer Jupyter Notebook
jupyter notebook

# Lancer JupyterLab
jupyter lab
```

### 4.3 – Anatomie d'un notebook Jupyter

Un notebook est composé de **cellules** de deux types :

- **Cellule Markdown** : texte formaté, titres, tableaux, formules LaTeX
- **Cellule Code** : code Python exécutable, avec affichage du résultat en dessous

**Raccourcis clavier essentiels :**

| Raccourci | Action |
|-----------|--------|
| `Shift + Enter` | Exécuter la cellule et passer à la suivante |
| `Ctrl + Enter` | Exécuter la cellule sans bouger |
| `A` (mode commande) | Insérer une cellule **au-dessus** |
| `B` (mode commande) | Insérer une cellule **en-dessous** |
| `M` (mode commande) | Convertir en cellule **Markdown** |
| `Y` (mode commande) | Convertir en cellule **Code** |
| `DD` (mode commande) | Supprimer la cellule |
| `Ctrl + Z` | Annuler dans une cellule |
| `Échap` | Passer en mode commande |
| `Entrée` | Passer en mode édition |

> ℹ️ **Mode commande** = cellule sélectionnée en bleu. **Mode édition** = curseur dans la cellule (bordure verte ou bleue selon le thème).

### 4.4 – Les commandes magiques IPython

Les **magic commands** sont des commandes spéciales préfixées par `%` (ligne) ou `%%` (cellule entière).

````python
# %timeit : mesurer le temps d'exécution d'une ligne
import numpy as np

%timeit sum(range(100_000))
````

````python
# Comparer Python pur vs NumPy
%timeit np.arange(100_000).sum()
````

````python
# %%timeit : mesurer le temps d'une cellule entière
%%timeit
L = list(range(100_000))
total = 0
for x in L:
    total += x
````

````python
# %who : lister les variables définies dans la session
a = 10
b = "bonjour"
c = [1, 2, 3]
%who
````

````python
# %whos : version plus détaillée avec types et valeurs
%whos
````

````python
# L'aide rapide avec ?
# Décommentez et exécutez pour voir la documentation
# np.array?

# Afficher l'aide d'une fonction Python
help(len)
````

## 📝 Partie 5 – Formats de fichiers scientifiques

En Data Science, vous rencontrerez de nombreux formats de fichiers. Voici un panorama :

### Formats tabulaires (données structurées)

| Format | Extension | Librairie Python | Avantages | Inconvénients |
|--------|-----------|-----------------|-----------|---------------|
| CSV | `.csv` | `pandas` | Universel, lisible | Pas de types, lent sur gros volumes |
| Excel | `.xlsx` | `pandas`, `openpyxl` | Très répandu en entreprise | Lourd, propriétaire |
| JSON | `.json` | `json`, `pandas` | Hiérarchique, web | Verbeux, peu optimisé |
| Parquet | `.parquet` | `pandas`, `pyarrow` | Très rapide, compressé | Binaire (non lisible) |
| HDF5 | `.h5`, `.hdf5` | `h5py`, `pandas` | Gros volumes, hiérarchique | Complexe à manipuler |

### Formats scientifiques spécialisés

| Format | Domaine | Librairie Python |
|--------|---------|----------------|
| NetCDF | Climatologie, météo, océanographie | `netCDF4`, `xarray` |
| GRIB | Prévisions météo | `cfgrib`, `eccodes` |
| MATLAB | Ingénierie | `scipy.io` |
| FITS | Astronomie | `astropy` |
| DICOM | Imagerie médicale | `pydicom` |

> 💡 **En pratique dans ce cours** : on travaillera principalement avec **CSV, Excel, JSON et HDF5/NetCDF** pour les données volumineuses.

## 📝 Partie 6 – Bonnes pratiques pour démarrer un projet Data

### Structure de projet recommandée

```
mon_projet/
├── data/
│   ├── raw/          ← données brutes (ne jamais modifier)
│   ├── processed/    ← données nettoyées
│   └── output/       ← résultats
├── notebooks/        ← vos notebooks Jupyter
├── src/              ← scripts Python réutilisables
├── environment.yml   ← définition de l'environnement conda
└── README.md         ← description du projet
```

### Règles d'or

1. **Un environnement par projet** — jamais installer dans `base`
2. **Versionner son code** — utiliser Git dès le début
3. **Ne jamais modifier les données brutes** — toujours travailler sur des copies
4. **Nommer clairement ses notebooks** — `01_exploration.ipynb`, `02_nettoyage.ipynb`…
5. **Documenter avec des cellules Markdown** — le notebook doit se lire comme un rapport
6. **Redémarrer et tout relancer** — vérifier que le notebook tourne du début à la fin sans erreur

> ⚠️ **Erreur fréquente** : exécuter les cellules dans le désordre. L'état du kernel dépend de l'ordre d'exécution, pas de l'ordre visuel des cellules. Toujours faire `Kernel > Restart & Run All` avant de livrer un notebook.
