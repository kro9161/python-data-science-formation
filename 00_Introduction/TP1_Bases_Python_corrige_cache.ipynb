## 📝 Partie 1 – Listes & boucles

### 🔎 Rappel théorique

Une **liste** est une collection **ordonnée** et **modifiable** d'éléments. Elle peut contenir des types différents, mais en pratique on y met souvent des données homogènes.

```python
ma_liste = [10, 20, 30, 40]   # création
ma_liste[0]                   # accès par index → 10
ma_liste[-1]                  # dernier élément → 40
ma_liste[1:3]                 # slice → [20, 30]
ma_liste.append(50)           # ajouter en fin
len(ma_liste)                 # longueur → 5
```

**Méthodes utiles d'une liste :**

| Méthode | Effet |
|---------|-------|
| `.append(x)` | Ajoute `x` en fin de liste |
| `.insert(i, x)` | Insère `x` à l'index `i` |
| `.remove(x)` | Supprime la première occurrence de `x` |
| `.pop(i)` | Supprime et retourne l'élément à l'index `i` |
| `.sort()` | Trie la liste sur place |
| `.reverse()` | Inverse la liste sur place |
| `.count(x)` | Compte les occurrences de `x` |
| `.index(x)` | Retourne l'index de la première occurrence de `x` |

````python
# Démonstration : liste de notes
notes = [12, 8, 15, 9, 14]
print("Notes :", notes)
print("Première note :", notes[0])
print("Dernière note :", notes[-1])
print("Notes de l'index 1 à 3 :", notes[1:3])

# Extraire les notes >= 10 avec une boucle classique
bonnes = []
for n in notes:
    if n >= 10:
        bonnes.append(n)
print("Notes >= 10 :", bonnes)
````

## 📝 Partie 2 – Tuples et Sets

### 🔎 Rappel théorique

#### Le tuple
Un **tuple** est une collection **ordonnée** et **immuable** (on ne peut pas la modifier après création).

```python
coordonnees = (48.8566, 2.3522)   # latitude, longitude de Paris
coordonnees[0]                    # → 48.8566
# coordonnees[0] = 0              # ❌ TypeError : un tuple ne se modifie pas
```

**Quand utiliser un tuple ?** Quand les données ne doivent pas changer : coordonnées GPS, couleurs RGB, dimensions d'une image, clés composées de dictionnaire.

#### Le set (ensemble)
Un **set** est une collection **non ordonnée** de valeurs **uniques** (pas de doublons).

```python
pays = {"France", "Allemagne", "France", "Espagne"}
print(pays)   # {'France', 'Allemagne', 'Espagne'}  ← doublon supprimé
```

| Opération | Code | Résultat |
|-----------|------|----------|
| Union | `A \| B` | Tous les éléments de A et B |
| Intersection | `A & B` | Éléments communs |
| Différence | `A - B` | Éléments dans A mais pas dans B |
| Appartenance | `x in A` | `True` ou `False` |

````python
# Démonstration : tuple
point = (3, 7)
print("Type :", type(point))
print("x =", point[0], "| y =", point[1])

# Démonstration : set
langages = {"Python", "Java", "Python", "C", "Java"}
print("\nLangages uniques :", langages)   # les doublons sont supprimés
print("Python est dans le set :", "Python" in langages)
````

## 📝 Partie 3 – Dictionnaires

### 🔎 Rappel théorique

Un **dictionnaire** est une collection de paires **clé → valeur**. Il est **non ordonné** (en Python < 3.7) mais **ordonné par insertion** depuis Python 3.7+.

```python
employe = {
    "nom": "Martin",
    "age": 35,
    "salaire": 3200
}

employe["nom"]           # accès → "Martin"
employe["poste"] = "DBA" # ajout d'une clé
employe.get("ville", "inconnue")  # accès sécurisé avec valeur par défaut
```

**Méthodes utiles :**

| Méthode | Retourne |
|---------|----------|
| `.keys()` | Toutes les clés |
| `.values()` | Toutes les valeurs |
| `.items()` | Paires (clé, valeur) |
| `.get(clé, défaut)` | Valeur ou défaut si clé absente |
| `clé in dict` | `True` si la clé existe |

````python
# Démonstration : dictionnaire de produits
produits = {"pizza": 12, "pâtes": 8, "salade": 6}
print("Produits :", produits)
print("Prix de la pizza :", produits["pizza"])

# Ajouter un produit
produits["dessert"] = 5
print("Après ajout :", produits)

# Itérer sur les paires clé/valeur
print("\nListe des produits :")
for nom, prix in produits.items():
    print(f"  {nom} : {prix} €")

# Calculer la moyenne des prix
total = 0
for prix in produits.values():
    total += prix
print("\nMoyenne des prix :", total / len(produits), "€")
````

## 📝 Partie 4 – Fonctions

### 🔎 Rappel théorique

Une **fonction** est un bloc de code réutilisable. Elle peut recevoir des **paramètres** et retourner une **valeur**.

```python
def nom_fonction(param1, param2):
    """Docstring : description de la fonction."""
    # corps de la fonction
    return resultat
```

**Bonnes pratiques :**
- Une fonction = **une seule responsabilité**
- Toujours écrire une **docstring** (description entre triple guillemets)
- Des **noms explicites** : `calculer_moyenne` plutôt que `calc` ou `f`
- **Valeurs par défaut** : `def saluer(nom, langue="fr")` → `langue` est optionnel

```python
# Exemple : paramètre avec valeur par défaut
def saluer(nom, langue="fr"):
    if langue == "fr":
        return "Bonjour " + nom
    else:
        return "Hello " + nom

saluer("Alice")          # → "Bonjour Alice"
saluer("Alice", "en")   # → "Hello Alice"
```

````python
# Démonstration : fonction moyenne avec docstring
def moyenne(liste):
    """Calcule la moyenne d'une liste de nombres.
    
    Paramètres:
        liste : list — liste de nombres
    Retourne:
        float — la moyenne
    """
    total = 0
    for valeur in liste:
        total += valeur
    return total / len(liste)

print(moyenne([10, 12, 14]))   # 12.0
print(moyenne([5, 7, 9, 11]))  # 8.0
````

## 📝 Partie 5 – Tout est objet en Python

### 🔎 Rappel théorique

En Python, **tout est objet** — sans exception. Un entier, une chaîne de caractères, une liste, une fonction, une classe… chaque valeur est une **instance** d'une classe.

Un objet possède :
- des **attributs** : données stockées dans l'objet (`ma_liste.__class__`)
- des **méthodes** : fonctions attachées à l'objet (`ma_liste.append()`)

```python
x = 42          # x est un objet de la classe int
type(x)         # → <class 'int'>
isinstance(x, int)  # → True
dir(x)          # → liste de tous les attributs et méthodes de x
```

**Pourquoi c'est important ?**  
Quand on fait `tableau.mean()` avec NumPy ou `df.groupby()` avec Pandas, on appelle des **méthodes d'objets**. Comprendre ce mécanisme est fondamental pour utiliser ces librairies efficacement.

### Aperçu des méthodes principales par type

| Type | Exemple de méthodes |
|------|--------------------|
| `str` | `.upper()`, `.lower()`, `.split()`, `.strip()`, `.replace()`, `.startswith()` |
| `list` | `.append()`, `.sort()`, `.reverse()`, `.count()`, `.index()` |
| `dict` | `.keys()`, `.values()`, `.items()`, `.get()` |
| `int` | `.bit_length()`, `.to_bytes()` |
| `float` | `.is_integer()`, `.as_integer_ratio()` |

````python
# Tout est objet : démonstration

# Un entier est un objet
x = 255
print("Type de x :", type(x))
print("x est un int ?", isinstance(x, int))
print("Nombre de bits nécessaires :", x.bit_length())   # 8 bits pour 255

# Une chaîne est un objet
texte = "  Python Data Science  "
print("\nType de texte :", type(texte))
print("Strip + lower :", texte.strip().lower())
print("Commence par 'Python' ?", texte.strip().startswith("Python"))
print("Nombre de mots :", len(texte.split()))

# Même une fonction est un objet
def dire_bonjour():
    return "Bonjour !"

print("\nType d'une fonction :", type(dire_bonjour))
print("Nom de la fonction :", dire_bonjour.__name__)
````

````python
# dir() : explorer les méthodes disponibles sur un objet
# (on filtre les méthodes 'dunder' __ pour la lisibilité)

s = "python"
methodes = [m for m in dir(s) if not m.startswith("_")]
print("Méthodes d'une chaîne str :")
print(methodes)
````
