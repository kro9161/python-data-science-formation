## 📝 Partie 1 – Le problème avec Python pur

### 🔎 Pourquoi Python seul ne suffit pas pour la data ?

Python est un langage **généraliste** : il peut tout faire, mais il n'est pas optimisé pour traiter de grandes quantités de nombres.

**Le problème :** quand vous créez une liste Python, Python emballe chaque nombre dans une "boîte" qui contient beaucoup d'informations en plus de la valeur (le type, des compteurs internes…). C'est pratique car ça permet de mélanger des types différents (`[1, "bonjour", True]`), mais ça prend beaucoup de place et c'est lent à traiter.

```
Liste Python [10, 20, 30] :

  ┌─────────────────┐    ┌─────────────────┐    ┌─────────────────┐
  │ Boîte n°1       │    │ Boîte n°2       │    │ Boîte n°3       │
  │ type : int      │    │ type : int      │    │ type : int      │
  │ valeur : 10     │    │ valeur : 20     │    │ valeur : 30     │
  │ (+ infos extra) │    │ (+ infos extra) │    │ (+ infos extra) │
  └─────────────────┘    └─────────────────┘    └─────────────────┘
     stockée quelque        stockée ailleurs        stockée ailleurs
     part en mémoire        en mémoire              en mémoire

Pour additionner 1 million de nombres → Python ouvre 1 million de boîtes
une par une → lent !
```

### 🔎 Ce que propose NumPy

NumPy stocke les nombres **côte à côte**, dans un bloc compact, **tous du même type** — plus de boîtes, plus d'infos en trop :

```
Tableau NumPy [10, 20, 30] :

  ┌────┬────┬────┐
  │ 10 │ 20 │ 30 │   ← tous au même endroit, même type, aucun emballage
  └────┴────┴────┘

Pour additionner 1 million de nombres → NumPy lit tout le bloc d'un coup
→ beaucoup plus rapide !
```

En plus, NumPy est écrit en **C** (un langage très rapide, proche de la machine) — les calculs ne passent pas par Python, ils sont exécutés directement par le processeur.

### 🔎 La vectorisation — plus de boucles !

Avec NumPy, on n'écrit plus de boucle `for` pour traiter chaque élément. On écrit une opération sur le tableau entier, et NumPy l'applique à **tous les éléments en même temps** :

```python
# Python pur — on traite chaque prix un par un avec une boucle
prix = [10, 20, 30]
prix_ttc = []
for p in prix:
    prix_ttc.append(p * 1.20)   # on répète cette ligne 3 fois

# NumPy — une seule instruction pour tout le tableau
prix = np.array([10, 20, 30])
prix_ttc = prix * 1.20    # NumPy applique ×1.20 à chaque élément automatiquement
```

C'est plus **court**, plus **lisible**, et beaucoup plus **rapide**.

## 📝 Partie 3 – Démonstration : Python pur vs NumPy

On mesure le temps nécessaire pour calculer la **somme de 1 million d'entiers**.

````python
import numpy as np
import time

N = 1_000_000

# --- Python pur : boucle explicite ---
L = list(range(N))
t1 = time.time()
total = 0
for x in L:
    total += x
t2 = time.time()
print(f"Python (boucle) : somme = {total}")
print(f"Temps             : {round(t2 - t1, 4)} secondes")

print()

# --- Python pur : sum() intégré (optimisé mais toujours Python) ---
t1 = time.time()
total2 = sum(L)
t2 = time.time()
print(f"Python (sum)    : somme = {total2}")
print(f"Temps             : {round(t2 - t1, 4)} secondes")

print()

# --- NumPy : vectorisé ---
A = np.arange(N)
t1 = time.time()
total3 = A.sum()
t2 = time.time()
print(f"NumPy           : somme = {total3}")
print(f"Temps             : {round(t2 - t1, 4)} secondes")
````

### Analyse des résultats

| Méthode | Temps typique | Facteur |
|---------|--------------|--------|
| Python boucle | ~50–100 ms | référence |
| Python `sum()` | ~5–15 ms | ~5–10x plus rapide |
| NumPy `.sum()` | ~0.5–2 ms | ~50–100x plus rapide |

> ⚠️ Les temps varient selon la machine. Ce qui compte : **l'ordre de grandeur**. NumPy est typiquement **10 à 100 fois plus rapide** que Python pur pour les opérations numériques.

### Démonstration : la vectorisation en pratique

Calcul du prix TTC pour une liste d'articles (TVA 20%).

````python
import numpy as np

prix_ht = np.array([10.0, 25.0, 50.0, 100.0, 200.0])
tva = 1.20

# Une seule ligne, aucune boucle — opération appliquée à tout le tableau
prix_ttc = prix_ht * tva

print("Prix HT  :", prix_ht)
print("Prix TTC :", prix_ttc)

# On peut aussi appliquer des opérations entre deux tableaux
remises = np.array([0.0, 0.05, 0.10, 0.15, 0.20])   # remises par article
prix_remise = prix_ht * (1 - remises)
print("Prix remisé :", prix_remise)
````

### Démonstration : occupation mémoire

````python
import sys
import numpy as np

N = 10_000

# Mémoire d'une liste Python
liste = list(range(N))
taille_liste = sys.getsizeof(liste) + sum(sys.getsizeof(x) for x in liste)

# Mémoire d'un tableau NumPy
tableau = np.arange(N, dtype=np.int64)
taille_numpy = tableau.nbytes

print(f"Liste Python  : {taille_liste:,} octets ({taille_liste // 1024} Ko)")
print(f"Tableau NumPy : {taille_numpy:,} octets ({taille_numpy // 1024} Ko)")
print(f"Ratio         : NumPy est ~{taille_liste // taille_numpy}x plus léger")
````

## ✅ Récapitulatif

| | Python pur | NumPy |
|-|-----------|-------|
| **Structure** | Liste d'objets hétérogènes | Bloc mémoire contigu et homogène |
| **Vitesse** | Lent (interprété, boucles) | Rapide (C/Fortran, vectorisé) |
| **Mémoire** | ~28 octets/entier | ~8 octets/int64 |
| **Syntaxe** | Boucles explicites | Opérations sur tout le tableau |
| **Typage** | Dynamique | Fixe (`dtype`) |

**3 notions clés à retenir :**
1. **Vectorisation** : opération sur tout un tableau en une ligne, sans boucle
2. **Homogénéité** : tous les éléments d'un tableau NumPy ont le même type (`dtype`)
3. **Performance** : 10 à 100x plus rapide que Python pur pour le calcul numérique
