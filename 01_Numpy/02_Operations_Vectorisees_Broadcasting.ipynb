## 📝 Partie 1 – Opérations vectorisées

### 🔎 Opérations élément par élément

Toutes les opérations arithmétiques de base (`+`, `-`, `*`, `/`, `**`, `%`) s'appliquent **élément par élément** sur les tableaux NumPy, sans boucle.

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

a + b    # → [11, 22, 33, 44]
a * b    # → [10, 40, 90, 160]
b / a    # → [10., 10., 10., 10.]
a ** 2   # → [1, 4, 9, 16]
```

**Opération scalaire** : si on combine un tableau avec un scalaire (un seul nombre), NumPy applique l'opération à chaque élément.

```python
a * 2    # → [2, 4, 6, 8]   — chaque élément × 2
a + 100  # → [101, 102, 103, 104]
```

### 🔎 `np.argmin` et `np.argmax`

Ces fonctions retournent l'**indice** (la position) de la valeur minimale ou maximale — pas la valeur elle-même.

```python
ventes = np.array([20, 35, 12, 48, 31])

np.max(ventes)      # → 48  (la valeur)
np.argmax(ventes)   # → 3   (l'indice de 48)

np.min(ventes)      # → 12
np.argmin(ventes)   # → 2   (l'indice de 12)
```

> 💡 **Cas d'usage** : "Quel jour a-t-on eu le plus de ventes ?" → `np.argmax(ventes)` donne le numéro du jour.

````python
import numpy as np

# Pizzeria 🍕 : ventes sur 5 jours
ventes = np.array([20, 22, 25, 30, 28])
prix_unitaire = 12.0

# Recettes journalières — opération scalaire
recettes = ventes * prix_unitaire
print("Recettes par jour (€) :", recettes)
print("Total semaine (€)     :", recettes.sum())

# Hausse de prix de 10 %
recettes_hausse = ventes * (prix_unitaire * 1.10)
print("Total après hausse 10% :", recettes_hausse.sum())

print()

# Opérations entre deux tableaux
ventes_s1 = np.array([20, 22, 25, 30, 28])
ventes_s2 = np.array([18, 25, 23, 35, 30])

evolution = ventes_s2 - ventes_s1          # évolution jour par jour
croissance = (ventes_s2 / ventes_s1) - 1   # taux de croissance
print("Évolution S2 vs S1 :", evolution)
print("Taux de croissance  :", np.round(croissance * 100, 1), "%")

print()

# argmin / argmax
print("Meilleur jour   : jour", np.argmax(ventes_s2), "avec", np.max(ventes_s2), "pizzas")
print("Moins bon jour  : jour", np.argmin(ventes_s2), "avec", np.min(ventes_s2), "pizzas")
````

## 📝 Partie 2 – Fonctions universelles (ufuncs)

### 🔎 Qu'est-ce qu'une ufunc ?

Une **ufunc** (*Universal Function*) est une fonction NumPy qui s'applique automatiquement à **chaque élément** d'un tableau, sans boucle. Vous passez un tableau, vous récupérez un tableau transformé.

C'est comme appuyer sur un bouton qui traite toutes vos données d'un coup.

### 🔎 Les ufuncs les plus utilisées en data

| Fonction | Ce qu'elle fait | Exemple |
|----------|----------------|--------|
| `np.sqrt(a)` | Racine carrée de chaque valeur | `np.sqrt([4, 9, 16])` → `[2. 3. 4.]` |
| `np.abs(a)` | Valeur absolue (supprime le signe moins) | `np.abs([-3, 1, -5])` → `[3 1 5]` |
| `np.round(a, n)` | Arrondit à n décimales | `np.round([1.567, 2.341], 2)` → `[1.57 2.34]` |
| `np.floor(a)` | Arrondit vers le bas (entier inférieur) | `np.floor([1.9, 2.1])` → `[1. 2.]` |
| `np.ceil(a)` | Arrondit vers le haut (entier supérieur) | `np.ceil([1.1, 2.9])` → `[2. 3.]` |
| `np.clip(a, min, max)` | Borne les valeurs entre un min et un max | `np.clip([0, 5, 15], 1, 10)` → `[1 5 10]` |
| `np.exp(a)` | Exponentielle — utilisée en ML et en stats | `np.exp([0, 1])` → `[1. 2.72]` |
| `np.log(a)` | Logarithme naturel — utilisé pour les rendements financiers | `np.log([1, np.e])` → `[0. 1.]` |

> 💡 **`np.clip`** est particulièrement pratique : il permet de "borner" les valeurs entre un minimum et un maximum. Utile pour corriger des données aberrantes sans les supprimer.
> ```python
> temperatures = np.array([-5, 10, 22, 50, 18])  # -5 et 50 sont suspects
> np.clip(temperatures, 0, 40)   # → [ 0 10 22 40 18]  (tout entre 0 et 40)
> ```
>
> 💡 **`np.exp` et `np.log`** : si vous ne connaissez pas ces fonctions mathématiques, ne vous inquiétez pas pour l'instant. Retenez juste que ce sont des fonctions qui existent et qui s'appliquent élément par élément comme toutes les autres.

````python
import numpy as np

# Météo 🌡️ : températures en °C sur 6 jours
temp_c = np.array([0, 10, 20, 30, 37, -5])

# Conversion °C → °F (formule vectorisée)
temp_f = (temp_c * 9/5) + 32
print("°C :", temp_c)
print("°F :", temp_f)

# Valeur absolue : utile pour les écarts
ecart_ref = temp_c - 20   # écart par rapport à 20°C
print("\nÉcart par rapport à 20°C :", ecart_ref)
print("Écart absolu              :", np.abs(ecart_ref))

# np.clip : on veut borner les températures entre -10 et 35
temp_bornee = np.clip(temp_c, -10, 35)
print("\nTempératures bornées [-10, 35] :", temp_bornee)

print()

# Finance : calcul de rendements logarithmiques (courant en finance)
cours = np.array([100.0, 102.5, 101.0, 105.0, 103.5])
rendements_log = np.log(cours[1:] / cours[:-1])
print("Cours         :", cours)
print("Rendements log:", np.round(rendements_log, 4))
````

## 📝 Partie 3 – L'axe (`axis`) dans les agrégations

### 🔎 Qu'est-ce qu'un axe ?

Pour un tableau 2D de shape `(lignes, colonnes)` :

```
         col0  col1  col2
ligne0 [  10    20    30  ]  ← axis=0 : direction "vers le bas" (par colonne)
ligne1 [  40    50    60  ]  ↓
ligne2 [  70    80    90  ]
                              axis=1 : direction "vers la droite" (par ligne) →
```

| `axis` | Direction | Résultat |
|--------|-----------|----------|
| `axis=0` | **Par colonne** (on "s'enfonce" dans les lignes) | 1 valeur par colonne |
| `axis=1` | **Par ligne** (on traverse les colonnes) | 1 valeur par ligne |
| *(aucun)* | **Global** (tout le tableau) | 1 seule valeur |

```python
A = np.array([[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]])

A.sum()         # → 450  (tout le tableau)
A.sum(axis=0)   # → [120, 150, 180]  (somme de chaque colonne)
A.sum(axis=1)   # → [60, 150, 240]   (somme de chaque ligne)
```

> 💡 **Moyen mnémotechnique** : `axis=0` → on réduit les **lignes** (on comprime verticalement).  
> `axis=1` → on réduit les **colonnes** (on comprime horizontalement).

````python
import numpy as np

# Netflix 🎬 : minutes vues — 7 jours × 3 séries
np.random.seed(0)
visionnages = np.random.randint(20, 60, (7, 3))
series = ["Série A", "Série B", "Série C"]

print("Visionnages (jours × séries) :\n", visionnages)
print("Shape :", visionnages.shape)   # (7, 3)

print()

# Total et moyenne PAR SÉRIE (axis=0 = on agrège les 7 lignes)
total_serie = visionnages.sum(axis=0)
moy_serie   = visionnages.mean(axis=0)
print("Total par série (axis=0) :", total_serie)
print("Moyenne par série        :", np.round(moy_serie, 1))
# → 1 valeur par colonne = 1 valeur par série

print()

# Total PAR JOUR (axis=1 = on agrège les 3 colonnes)
total_jour = visionnages.sum(axis=1)
print("Total par jour (axis=1) :", total_jour)
# → 1 valeur par ligne = 1 valeur par jour

# Quel jour a-t-on le plus regardé ?
jour_max = np.argmax(total_jour)
print("\nJour le plus regardé : jour", jour_max,
      "avec", total_jour[jour_max], "minutes au total")

# Quelle série est la plus populaire ?
serie_max = np.argmax(moy_serie)
print("Série la plus populaire :", series[serie_max],
      "avec", round(moy_serie[serie_max], 1), "min/jour en moyenne")
````

## 📝 Partie 4 – Broadcasting

### 🔎 Le problème que le broadcasting résout

Imaginez : vous avez un tableau de ventes de 4 articles sur 3 jours (shape `(3, 4)`), et vous voulez appliquer une remise **différente** à chaque article.

```
Ventes (3 jours × 4 articles) :    Remise par article :
[[100, 200, 300, 400],              [0.10, 0.20, 0.00, 0.15]
 [110, 210, 310, 410],              (remises différentes pour chaque article)
 [120, 220, 320, 420]]
```

Intuitivement, on veut multiplier **chaque colonne** par la remise correspondante. Mais `ventes.shape = (3, 4)` et `remise.shape = (4,)` — les shapes ne sont pas identiques.

NumPy résout ça automatiquement : il "étire" le vecteur `remise` pour qu'il s'applique à chaque ligne :

```
remise = [0.10, 0.20, 0.00, 0.15]   shape: (4,)

NumPy l'étire mentalement comme ça :
[[0.10, 0.20, 0.00, 0.15],          shape: (3, 4)
 [0.10, 0.20, 0.00, 0.15],
 [0.10, 0.20, 0.00, 0.15]]

→ ventes * remise fonctionne ! C'est le broadcasting.
```

### 🔎 La règle simple du broadcasting

NumPy peut faire le broadcasting entre deux tableaux **si leurs shapes sont compatibles**. Pour savoir si c'est le cas, on compare les dimensions **de droite à gauche** :
- Si deux dimensions sont **égales** → OK
- Si l'une d'elles est **1** → OK (NumPy répète cette dimension)
- Sinon → erreur

```
Cas 1 : (3, 4) et (4,)   → OK ✅
  de droite à gauche : 4 == 4  ✓  |  3 vs rien → NumPy préfixe par 1 → 1 répété 3 fois ✓

Cas 2 : (3, 4) et (3, 1) → OK ✅
  de droite à gauche : 4 vs 1 → la colonne unique est répétée 4 fois ✓  |  3 == 3 ✓

Cas 3 : (3, 4) et (3,)   → ERREUR ❌
  de droite à gauche : 4 vs 3 → ni égaux ni l'un des deux = 1 → impossible
```

### 🔎 Le piège le plus fréquent

Vous avez un vecteur de 3 valeurs et un tableau de 3 lignes. Vous voulez appliquer une valeur différente **à chaque ligne**.

```python
A = np.ones((3, 4))          # 3 lignes, 4 colonnes
b = np.array([10, 20, 30])   # 3 valeurs, une par ligne

# A * b  → ERREUR ! shape (3,4) vs shape (3,) ne matche pas
# NumPy compare : 4 vs 3 → incompatible ❌

# Solution : transformer b en colonne → shape (3, 1)
b_col = b.reshape(-1, 1)     # shape (3, 1)
A * b_col                    # → OK ! 1 est répété 4 fois → (3, 4) ✅
```

**Visualisation :**
```
b_col = [[10],    →  étiré en  →  [[10, 10, 10, 10],
          [20],                     [20, 20, 20, 20],
          [30]]                     [30, 30, 30, 30]]

→ chaque ligne de A est multipliée par la valeur correspondante de b_col
```

> 💡 **La règle à retenir :** si vous voulez une valeur différente par **colonne** (par article), votre vecteur doit avoir la shape `(n,)` ou `(1, n)`.  
> Si vous voulez une valeur différente par **ligne** (par client), votre vecteur doit avoir la shape `(n, 1)` — utilisez `.reshape(-1, 1)`.

````python
import numpy as np

# E-commerce 🛍️ : prix HT de 4 articles sur 3 catégories
# Shape : (4, 3)  → 4 articles × 3 catégories
prix_ht = np.array([[100, 50,  30],
                    [200, 80,  45],
                    [150, 60,  25],
                    [120, 90,  55]])

# --- Broadcasting sur les colonnes : remise par catégorie ---
# remise_cat.shape = (3,) → compatible avec (4, 3) car 3 == 3
remise_cat = np.array([0.10, 0.05, 0.20])   # -10%, -5%, -20%
prix_remise = prix_ht * (1 - remise_cat)
print("Prix après remise par catégorie :\n", prix_remise)
print("(shape du résultat :", prix_remise.shape, ")")

print()

# --- Broadcasting sur les lignes : coeff par article ---
# coeff_art.shape = (4,) → INCOMPATIBLE avec (4, 3) !
# Il faut (4, 1) pour broadcaster sur les colonnes
coeff_art = np.array([1.0, 1.2, 0.9, 1.1]).reshape(-1, 1)  # → shape (4, 1)
print("Shape de coeff_art :", coeff_art.shape)              # (4, 1)
prix_pondere = prix_ht * coeff_art
print("Prix pondéré par article :\n", prix_pondere)

print()

# --- TVA : scalaire → compatible avec tout ---
tva = 1.20
prix_ttc = prix_remise * tva
print("Prix TTC :\n", np.round(prix_ttc, 2))
````

````python
import numpy as np

# Démonstration du piège : (4, 3) × (4,) → ValueError
A = np.ones((4, 3))
b = np.array([1, 2, 3, 4])   # shape (4,) — 4 éléments

print("A.shape :", A.shape)   # (4, 3)
print("b.shape :", b.shape)   # (4,)

# Décommentez pour voir l'erreur :
# A * b   # → ValueError: operands could not be broadcast with shapes (4,3) (4,)

# Solution : transformer b en colonne (4, 1)
b_col = b.reshape(-1, 1)    # ou b[:, np.newaxis]
print("b_col.shape :", b_col.shape)   # (4, 1)
print("A * b_col :\n", A * b_col)     # fonctionne ✅
````

| **Vectorisation** | `+`, `-`, `*`, `/`, `**` s'appliquent élément par élément — pas besoin de boucle |
| **Scalaire** | Un seul nombre s'applique à tout le tableau automatiquement |
| **`np.argmax/argmin`** | Retourne l'**indice** du max/min (pas la valeur) |
| **Ufuncs** | `np.sqrt`, `np.abs`, `np.exp`, `np.log`, `np.clip`… — opèrent élément par élément |
| **`axis=0`** | Agrégation **par colonne** (on réduit les lignes) → 1 résultat par colonne |
| **`axis=1`** | Agrégation **par ligne** (on réduit les colonnes) → 1 résultat par ligne |
| **Broadcasting** | NumPy étend automatiquement les petits tableaux pour les rendre compatibles |
| **Règle broadcasting** | Les shapes sont comparées de droite à gauche — dimensions égales ou égale à 1 |
| **Piège broadcasting** | Un vecteur `(n,)` ne se broadcast pas sur les lignes d'une matrice `(n, m)` — utiliser `reshape(-1, 1)` |
