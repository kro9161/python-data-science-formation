## 📝 Partie 1 – Une image, c'est quoi pour NumPy ?

### 🔎 Un pixel, c'est quoi ?

Un écran est composé de milliers de petits points lumineux appelés **pixels** (picture elements). Chaque pixel a une couleur. Pour décrire cette couleur, on utilise trois valeurs : la quantité de **Rouge**, de **Vert** et de **Bleu** — c'est le système **RGB** (Red, Green, Blue).

```
Chaque valeur est un entier entre 0 et 255 :
  0   = couleur absente
  255 = couleur à fond

Exemples :
  Rouge pur    → R=255, G=0,   B=0
  Vert pur     → R=0,   G=255, B=0
  Bleu pur     → R=0,   G=0,   B=255
  Blanc        → R=255, G=255, B=255  (toutes les couleurs à fond)
  Noir         → R=0,   G=0,   B=0   (toutes les couleurs absentes)
  Gris moyen   → R=128, G=128, B=128
  Jaune        → R=255, G=255, B=0   (rouge + vert = jaune)
```

### 🔎 Une image = un tableau NumPy 3D

Une image de 858 pixels de hauteur × 1280 pixels de largeur en couleur est stockée comme un tableau NumPy de shape **(858, 1280, 3)** :

```
img.shape = (858, 1280, 3)
              ↑     ↑    ↑
           hauteur  │   3 canaux : Rouge, Vert, Bleu
                 largeur

Pour accéder à un pixel précis :
  img[100, 200]        → les 3 valeurs RGB du pixel à la ligne 100, colonne 200
                          ex: [245, 180, 60]  (un orange)

  img[100, 200, 0]     → uniquement le canal Rouge de ce pixel → 245
  img[100, 200, 1]     → uniquement le canal Vert              → 180
  img[100, 200, 2]     → uniquement le canal Bleu              → 60

Pour extraire un canal entier :
  img[:, :, 0]         → tous les pixels, canal Rouge seulement (tableau 2D)
  img[:, :, 1]         → tous les pixels, canal Vert
  img[:, :, 2]         → tous les pixels, canal Bleu
```

## 📝 Partie 2 – Zoom et découpe (slicing visuel)

### 🔎 Comment découper une image ?

Découper une image, c'est exactement la même chose que faire un slicing sur un tableau NumPy 2D. On choisit une **plage de lignes** (hauteur) et une **plage de colonnes** (largeur).

```
img[ligne_debut:ligne_fin, colonne_debut:colonne_fin]

Exemple : extraire un carré de 150×150 pixels à partir du coin (100, 150)
  img[100:250, 150:300]
       ↑    ↑   ↑    ↑
    ligne  ligne col  col
    début  fin  début fin
```

**Visualisation :**
```
Image complète (858 × 1280 px) :

  0 ─────────────────────────── 1280
  │                              │
  │    ┌──────────┐              │
 100   │  ZOOM    │              │
  │    │ 150×150  │              │
 250   └──────────┘              │
  │   150        300             │
 858 ─────────────────────────── │
```

> 💡 Le slicing retourne une **vue** sur le tableau original (pas une copie). Si on modifie le zoom, ça modifie l'image originale. Pour travailler indépendamment, utilisez `.copy()`.

````python
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('images/meteo.jpg')

# Zoom sur le centre de la carte météo
h, w = img.shape[0], img.shape[1]
cx, cy = h // 2, w // 2   # centre de l'image
taille = 150

zoom_centre = img[cx-taille:cx+taille, cy-taille:cy+taille]
print(f"Shape du zoom : {zoom_centre.shape}")   # (300, 300, 3)

# Zoom sur un coin précis (haut gauche)
zoom_coin = img[50:250, 50:300]
print(f"Shape du coin : {zoom_coin.shape}")

# Afficher côte à côte
fig, axes = plt.subplots(1, 3, figsize=(14, 5))
axes[0].imshow(img);           axes[0].set_title('Image originale'); axes[0].axis('off')
axes[1].imshow(zoom_centre);   axes[1].set_title('Zoom centre');     axes[1].axis('off')
axes[2].imshow(zoom_coin);     axes[2].set_title('Zoom coin');       axes[2].axis('off')
plt.tight_layout()
plt.show()
````

## 📝 Partie 3 – Symétries

### 🔎 Comment retourner une image ?

Retourner une image, c'est inverser l'ordre des lignes ou des colonnes. En NumPy, `[::-1]` inverse l'ordre des éléments — on l'a vu sur les tableaux 1D. Sur une image, ça marche pareil mais on choisit quelle dimension inverser.

```
Image originale :
  ligne 0  → [pixel 0, pixel 1, pixel 2, ...]
  ligne 1  → [pixel 0, pixel 1, pixel 2, ...]
  ligne 2  → [pixel 0, pixel 1, pixel 2, ...]

Symétrie VERTICALE (miroir haut/bas) : img[::-1, :]
  → on inverse l'ordre des LIGNES
  ligne 0 devient la dernière, la dernière devient ligne 0
  → l'image est retournée comme si on la regardait à l'envers

Symétrie HORIZONTALE (miroir gauche/droite) : img[:, ::-1]
  → on inverse l'ordre des COLONNES dans chaque ligne
  → l'image est retournée comme dans un miroir
```

> 💡 Le `:` seul signifie "tout garder". Donc `img[::-1, :]` = "toutes les lignes inversées, toutes les colonnes inchangées".

````python
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('images/netflix.jpg')

# Symétrie horizontale (miroir gauche/droite)
sym_h = img[:, ::-1]
print("Shape sym_h :", sym_h.shape)   # identique à img.shape

# Symétrie verticale (miroir haut/bas)
sym_v = img[::-1, :]
print("Shape sym_v :", sym_v.shape)

# Double symétrie (rotation 180°)
sym_hv = img[::-1, ::-1]

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(img);      axes[0].set_title('Originale');         axes[0].axis('off')
axes[1].imshow(sym_h);    axes[1].set_title('Miroir G/D');        axes[1].axis('off')
axes[2].imshow(sym_v);    axes[2].set_title('Miroir H/B');        axes[2].axis('off')
axes[3].imshow(sym_hv);   axes[3].set_title('Double miroir');     axes[3].axis('off')
plt.suptitle('Symétries', fontsize=13)
plt.tight_layout()
plt.show()
````

## 📝 Partie 4 – Niveaux de gris

### 🔎 Comment passer d'une image couleur à une image en gris ?

Une image couleur a 3 canaux (R, G, B). Une image en niveaux de gris n'en a qu'**un seul** : l'intensité lumineuse.

**L'idée :** pour chaque pixel, on calcule la **moyenne** des trois valeurs R, G, B. Cette moyenne représente la luminosité du pixel.

```
Pixel original : R=200, G=150, B=100
Moyenne        : (200 + 150 + 100) / 3 = 150
→ Ce pixel devient gris à 150 (sur 255)

Pixel blanc : R=255, G=255, B=255
Moyenne     : (255 + 255 + 255) / 3 = 255  → reste blanc

Pixel noir  : R=0, G=0, B=0
Moyenne     : (0 + 0 + 0) / 3 = 0          → reste noir
```

En NumPy, on utilise `.mean(axis=2)` : on calcule la moyenne sur **l'axe 2** (le 3e axe = les 3 canaux).

```python
img.shape          # (858, 1280, 3)  → tableau 3D
img.mean(axis=2)   # (858, 1280)     → tableau 2D : la 3e dimension a été réduite à sa moyenne
```

> ⚠️ La méthode `.mean(axis=2)` est une approximation rapide. En réalité, l'œil humain est plus sensible au vert qu'au rouge et au bleu. La formule plus précise est : `0.299 × R + 0.587 × G + 0.114 × B`. Mais pour comprendre le principe, la moyenne suffit.

````python
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('images/meteo.jpg')

# Méthode 1 : moyenne simple des 3 canaux
gris_simple = img.mean(axis=2)
print("Shape couleur :", img.shape)         # (858, 1280, 3)
print("Shape gris    :", gris_simple.shape) # (858, 1280)  — plus de 3e dimension
print("dtype gris    :", gris_simple.dtype) # float64

# Méthode 2 : formule perceptuelle (plus précise)
gris_perc = 0.299 * img[:,:,0] + 0.587 * img[:,:,1] + 0.114 * img[:,:,2]

# Afficher les deux
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
axes[0].imshow(img);           axes[0].set_title('Couleur (RGB)');      axes[0].axis('off')
axes[1].imshow(gris_simple, cmap='gray'); axes[1].set_title('Gris (moyenne)'); axes[1].axis('off')
axes[2].imshow(gris_perc, cmap='gray');  axes[2].set_title('Gris (perceptuel)'); axes[2].axis('off')
plt.tight_layout()
plt.show()
````

## 📝 Partie 5 – Inversion des couleurs

### 🔎 Comment inverser les couleurs ?

Inverser les couleurs d'une image, c'est **retourner chaque valeur par rapport à son maximum**. Un pixel très clair (proche de 255) devient très sombre (proche de 0), et inversement.

```
Si les pixels sont entre 0 et 255 (entiers) :
  pixel inversé = 255 - pixel original

  Pixel blanc  : 255 → 255 - 255 = 0   → devient noir
  Pixel noir   :   0 → 255 - 0   = 255 → devient blanc
  Pixel gris   : 128 → 255 - 128 = 127 → reste gris (symétrique)
  Rouge pur    : R=255,G=0,B=0 → R=0,G=255,B=255 → devient cyan

Si les pixels sont entre 0 et 1 (flottants, après normalisation) :
  pixel inversé = 1 - pixel original
```

C'est une opération vectorisée NumPy : elle s'applique à tous les pixels en une seule instruction, sans boucle.

````python
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('images/netflix.jpg')

# Vérifier le type de valeurs
print("dtype :", img.dtype)
print("max   :", img.max())

# Inversion (s'adapte automatiquement selon le type)
if img.max() > 1:
    inverse = 255 - img    # valeurs entières 0-255
else:
    inverse = 1 - img      # valeurs flottantes 0-1

# Vérification
print(f"\nPixel original  : {img[100, 200]}")
print(f"Pixel inversé   : {inverse[100, 200]}")
print(f"Somme (doit = 255) : {img[100, 200] + inverse[100, 200]}")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img);      axes[0].set_title('Originale');  axes[0].axis('off')
axes[1].imshow(inverse);  axes[1].set_title('Inversée');   axes[1].axis('off')
plt.tight_layout()
plt.show()
````

## 📝 Partie 6 – Normalisation des pixels

### 🔎 Pourquoi normaliser une image ?

Les valeurs brutes des pixels sont des entiers entre **0 et 255**. Pour beaucoup d'algorithmes de traitement d'image et surtout pour les modèles d'**intelligence artificielle**, on a besoin de valeurs entre **0 et 1**.

**Pourquoi ?** Les algorithmes de ML ont du mal avec de grandes valeurs (0-255). En ramenant tout entre 0 et 1, les calculs sont plus stables et les modèles convergent mieux.

**La formule :** on divise simplement chaque pixel par 255 (la valeur maximale possible).

```
Avant normalisation :
  pixel noir   =   0  → 0 / 255 = 0.0
  pixel gris   = 128  → 128 / 255 ≈ 0.502
  pixel blanc  = 255  → 255 / 255 = 1.0

Après normalisation : toutes les valeurs sont entre 0.0 et 1.0
```

**Le dtype change aussi :** en divisant par 255.0 (un flottant), NumPy convertit automatiquement le tableau de `uint8` (entiers 0-255) à `float64` (flottants).

```python
img.dtype        # uint8
(img / 255.0).dtype  # float64
```

````python
import numpy as np
import matplotlib.pyplot as plt
img = plt.imread('images/pizza.jpg')

print("=== Avant normalisation ===")
print(f"dtype : {img.dtype}")
print(f"min   : {img.min()}  | max : {img.max()}")
print(f"Pixel (100,200) : {img[100, 200]}")

# Normalisation
img_norm = img / 255.0

print("\n=== Après normalisation ===")
print(f"dtype : {img_norm.dtype}")
print(f"min   : {img_norm.min():.4f}  | max : {img_norm.max():.4f}")
print(f"Pixel (100,200) : {np.round(img_norm[100, 200], 3)}")

# Les deux images sont visuellement identiques
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img);       axes[0].set_title('Original (uint8, 0-255)');   axes[0].axis('off')
axes[1].imshow(img_norm);  axes[1].set_title('Normalisé (float64, 0-1)');  axes[1].axis('off')
plt.tight_layout()
plt.show()
print("\n→ Visuellement identiques, mais les valeurs numériques sont différentes.")
````

## 🤖 Lien avec l'intelligence artificielle

### Pourquoi tout ça est fondamental pour l'IA ?

Un modèle d'intelligence artificielle ne **voit pas** une image comme nous. Il ne distingue pas une pizza d'un chat. Il lit uniquement **des nombres** — exactement les valeurs de pixels qu'on vient de manipuler.

Avant d'entraîner un modèle sur des images, on fait **toujours** les mêmes étapes de préparation :

```
1. Charger l'image          → img = plt.imread(...)       tableau NumPy (H, W, 3)
2. Redimensionner           → img_resize = ...            taille fixe ex: (224, 224, 3)
3. Normaliser               → img / 255.0                 valeurs entre 0 et 1
4. Convertir en float32     → .astype('float32')          format attendu par les modèles
5. Créer un batch           → img[None, ...]              ajouter une dim pour le lot
                              → (1, 224, 224, 3)          1 image dans le batch
```

**Exemple de batch** :

```python
# Préparer une image pour un modèle (ex: ResNet, VGG, YOLO)
img_norm  = (img / 255.0).astype('float32')  # normaliser
img_batch = img_norm[None, ...]              # shape : (1, H, W, 3)
                                             # le 1 = "un seul exemple dans le batch"

# Si on avait 32 images : shape = (32, H, W, 3)
```

**Ce qu'on a appris dans ce chapitre, c'est exactement ce que font les frameworks d'IA (TensorFlow, PyTorch) en coulisse.** La seule différence : ils font ça sur des milliers d'images à la fois et ajoutent des transformations aléatoires (rotations, recadrages, changements de luminosité) pour rendre les modèles plus robustes — c'est ce qu'on appelle la **data augmentation**.

````python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('images/pizza.jpg')

# Pipeline de préparation d'image pour un modèle IA
print("=== Pipeline de préparation IA ===")
print(f"1. Image brute          : {img.shape} | dtype={img.dtype} | max={img.max()}")

img_norm = img / 255.0
print(f"2. Après normalisation  : {img_norm.shape} | dtype={img_norm.dtype} | max={img_norm.max():.2f}")

img_f32 = img_norm.astype('float32')
print(f"3. Après float32        : {img_f32.shape} | dtype={img_f32.dtype}")

img_batch = img_f32[None, ...]   # ajoute une dimension en position 0
print(f"4. Après batch (None)   : {img_batch.shape}")
print(f"   → On lit : 1 image, {img_batch.shape[1]}px de haut, {img_batch.shape[2]}px de large, 3 canaux")

print("\n✅ L'image est prête à être passée à un modèle d'IA !")
````

## ✅ Récapitulatif

| Concept | Ce qu'il faut retenir |
|---------|----------------------|
| **Pixel** | Point élémentaire d'une image, décrit par 3 valeurs : R, G, B (0-255) |
| **Shape image** | `(hauteur, largeur, 3)` — tableau NumPy 3D |
| **Canaux** | `img[:,:,0]` = Rouge, `img[:,:,1]` = Vert, `img[:,:,2]` = Bleu |
| **Zoom / découpe** | Slicing classique : `img[100:250, 150:300]` |
| **Miroir gauche/droite** | `img[:, ::-1]` — on inverse l'ordre des colonnes |
| **Miroir haut/bas** | `img[::-1, :]` — on inverse l'ordre des lignes |
| **Niveaux de gris** | `img.mean(axis=2)` — on moyenne les 3 canaux → shape passe de (H,W,3) à (H,W) |
| **Inversion couleurs** | `255 - img` (entiers) ou `1 - img` (flottants) |
| **Normalisation** | `img / 255.0` — passe de 0-255 à 0.0-1.0, dtype passe de uint8 à float64 |
| **Batch IA** | `img[None, ...]` — ajoute une dimension pour le lot d'images |
