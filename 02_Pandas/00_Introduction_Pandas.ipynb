## 📝 Partie 1 – Pourquoi Pandas ?

### 🔎 Le problème sans Pandas

Imaginez un fichier Excel avec 50 000 lignes de ventes. Vous voulez :
- Calculer le chiffre d'affaires moyen par région
- Filtrer uniquement les ventes du mois de mars
- Fusionner avec un autre fichier de clients

Avec Excel, c'est faisable mais **lent, fragile, et difficile à automatiser**. Avec Python pur (listes, dictionnaires), c'est possible mais il faut écrire des dizaines de lignes de boucles.

**Pandas résout ça** : c'est une bibliothèque Python spécialement conçue pour manipuler des données tabulaires (tableaux avec lignes et colonnes), exactement comme Excel mais **en code**, ce qui permet de :
- Traiter des millions de lignes rapidement
- Automatiser les analyses (même traitement, nouvelles données → relancer le script)
- Reproduire exactement les mêmes calculs sans risque d'erreur humaine

### 🔎 Pandas vs Excel — à quoi ça correspond ?

| Excel | Pandas | Description |
|-------|--------|-------------|
| Classeur | — | Le fichier entier |
| Feuille | `DataFrame` | Un tableau avec lignes et colonnes |
| Colonne | `Series` | Une seule colonne de données |
| Cellule | `df.loc[ligne, colonne]` | Une valeur précise |
| `=MOYENNE()` | `.mean()` | Calcul de la moyenne |
| `=MIN()` / `=MAX()` | `.min()` / `.max()` | Minimum / Maximum |
| Filtre automatique | `df[df["col"] > valeur]` | Filtrer les lignes |
| Tableau croisé dynamique | `.groupby()` / `.pivot_table()` | Regrouper et agréger |
| Enregistrer sous... | `.to_csv()` / `.to_excel()` | Exporter les données |

## 📝 Partie 2 – La Series : une colonne intelligente

### 🔎 Qu'est-ce qu'une Series ?

Une **Series**, c'est l'équivalent d'**une colonne Excel** — mais avec des super-pouvoirs.

Concrètement :
- Elle contient une liste de valeurs (nombres, textes, dates…)
- Chaque valeur a une **étiquette** qu'on appelle l'**index** (par défaut 0, 1, 2…)
- Elle connaît le **type** de ses données
- Elle peut faire des **calculs** directement (moyenne, tri, filtre…)

```
Colonne Excel "Âge" :        Series Pandas "ages" :

┌───┬─────┐                  index  valeur
│ 1 │  25 │                    0  →   25
│ 2 │  32 │                    1  →   32
│ 3 │  41 │                    2  →   41
│ 4 │  29 │                    3  →   29
│ 5 │  35 │                    4  →   35
└───┴─────┘                  dtype: int64

La différence : Pandas connaît le type (int64)
et peut faire des calculs dessus directement.
```

### 🔎 L'index — à quoi ça sert ?

L'**index** est le numéro de chaque ligne. Par défaut, Pandas numérote à partir de 0 (comme Python). C'est ce qui permet d'accéder à une valeur précise :

```python
ages[0]   # → 25  (la première valeur)
ages[2]   # → 41  (la troisième valeur)
```

On peut aussi donner des **étiquettes personnalisées** à l'index :
```python
ages = pd.Series([25, 32, 41], index=["Alice", "Bob", "Chloé"])
ages["Bob"]   # → 32
```

````python
import pandas as pd

# Créer une Series à partir d'une liste Python
ages = pd.Series([25, 32, 41, 29, 35])

print("La Series :")
print(ages)
print()
print("Type de l'objet  :", type(ages))    # pd.Series
print("Type des données :", ages.dtype)     # int64
print("Nombre de valeurs:", len(ages))      # 5
print()
print("Accès par index  :", ages[0])        # première valeur → 25
print("Accès par index  :", ages[2])        # troisième valeur → 41
````

````python
# Series avec index personnalisé (étiquettes au lieu de 0,1,2…)
salaires = pd.Series(
    [2500, 3200, 4100, 2900],
    index=["Alice", "Bob", "Chloé", "David"]
)

print(salaires)
print()
print("Salaire de Bob   :", salaires["Bob"])    # accès par étiquette
print("Salaire de Chloé :", salaires["Chloé"])
````

## 📝 Partie 3 – Calculs sur une Series

### 🔎 Les méthodes de calcul essentielles

Une fois la Series créée, Pandas offre directement des méthodes pour l'analyser — sans boucle, sans formule complexe.

| Méthode Pandas | Équivalent Excel | Ce qu'elle calcule |
|---------------|-----------------|-------------------|
| `.mean()` | `=MOYENNE(A1:A5)` | La moyenne des valeurs |
| `.median()` | `=MEDIANE(A1:A5)` | La médiane |
| `.min()` | `=MIN(A1:A5)` | La valeur minimale |
| `.max()` | `=MAX(A1:A5)` | La valeur maximale |
| `.sum()` | `=SOMME(A1:A5)` | La somme totale |
| `.std()` | `=ECARTYPE(A1:A5)` | L'écart-type |
| `.count()` | `=NB(A1:A5)` | Le nombre de valeurs non vides |
| `.sort_values()` | Trier A→Z | Trier les valeurs |
| `.value_counts()` | Tableau croisé simple | Compter les occurrences |
| `.describe()` | Statistiques descriptives | Résumé complet en une ligne |

````python
import pandas as pd

ages = pd.Series([25, 32, 41, 29, 35], name="Âge")

print(f"Moyenne   : {ages.mean():.1f} ans")
print(f"Médiane   : {ages.median():.1f} ans")
print(f"Minimum   : {ages.min()} ans")
print(f"Maximum   : {ages.max()} ans")
print(f"Somme     : {ages.sum()} ans (total)")
print(f"Écart-type: {ages.std():.2f}")
````

````python
# describe() : le résumé statistique complet en une seule ligne
# C'est l'équivalent de sélectionner une colonne Excel et regarder
# la barre de statut en bas (Moyenne, Min, Max...)
# mais en beaucoup plus complet !

print(ages.describe())
````

### 🔎 Lire le résultat de `describe()`

```
count     5.0    → nombre de valeurs (5 personnes)
mean     32.4    → moyenne des âges
std       6.07   → écart-type : les âges varient de ±6 ans autour de la moyenne
min      25.0    → le plus jeune a 25 ans
25%      29.0    → 1 personne sur 4 a moins de 29 ans
50%      32.0    → la moitié a moins de 32 ans (= médiane)
75%      35.0    → 3 personnes sur 4 ont moins de 35 ans
max      41.0    → le plus âgé a 41 ans
```

> 💡 **Astuce :** `describe()` est souvent la **première chose qu'on fait** quand on reçoit un nouveau jeu de données. En une ligne, on comprend l'étendue, la dispersion et la distribution des données.

````python
import pandas as pd

# sort_values() : trier les valeurs
ages = pd.Series([25, 32, 41, 29, 35], name="Âge")

print("Tri croissant (par défaut) :")
print(ages.sort_values())
print()
print("Tri décroissant :")
print(ages.sort_values(ascending=False))
````

````python
# value_counts() : compter les occurrences — utile sur du texte
# Exemple : combien d'employés par département ?

departements = pd.Series(["IT", "RH", "IT", "Marketing", "IT", "RH", "Finance"])

print("Nombre d'employés par département :")
print(departements.value_counts())
# → IT         3
# → RH         2
# → Marketing  1
# → Finance    1
````

## 📝 Partie 4 – Filtrer une Series

### 🔎 Sélectionner des valeurs selon une condition

Comme avec NumPy, on peut filtrer une Series avec une condition booléenne. C'est l'équivalent du **filtre automatique** dans Excel.

```python
ages = pd.Series([25, 32, 41, 29, 35])

# Garder uniquement les âges > 30
ages[ages > 30]   # → [32, 41, 35]

# Combien de personnes ont plus de 30 ans ?
(ages > 30).sum()   # → 3
```

````python
import pandas as pd

salaires = pd.Series(
    [2500, 3200, 4100, 2900, 5000, 2200],
    index=["Alice", "Bob", "Chloé", "David", "Eva", "Fabien"]
)

# Employés avec un salaire > 3000 €
print("Salaires > 3000 € :")
print(salaires[salaires > 3000])

print()
print(f"Nombre d'employés > 3000 € : {(salaires > 3000).sum()}")
print(f"Salaire moyen des > 3000 € : {salaires[salaires > 3000].mean():.0f} €")
````
