## 📝 Partie 3 – Filtrer les données

### 🔎 Rappel : filtrer avec une condition

```python
df[df["Note"] >= 8]         # Œuvres bien notées
df[df["Genre"] == "Film"]   # Uniquement les films
df[df["Année"] > 2018]      # Sortis après 2018
```

### 🔎 Combiner plusieurs conditions : `&` (ET) et `|` (OU)

Pour combiner plusieurs conditions, on utilise `&` (ET) ou `|` (OU) — **jamais** `and` ou `or`.

```python
# ET logique : les deux conditions doivent être vraies
df[(df["Genre"] == "Film") & (df["Note"] >= 8)]
#   ↑ film                    ↑ ET bien noté

# OU logique : au moins une des conditions est vraie
df[(df["Genre"] == "Film") | (df["Année"] >= 2020)]
#   ↑ film                  ↑ OU sorti en 2020 ou après
```

> ⚠️ **Les parenthèses sont obligatoires** autour de chaque condition. Sans elles, Python interprète mal la priorité des opérateurs et génère une erreur.

### 🔎 `value_counts()` — compter les occurrences

`value_counts()` compte combien de fois chaque valeur apparaît dans une colonne. C'est l'équivalent d'un `NBSI()` dans Excel ou d'un `COUNT(*)... GROUP BY` en SQL.

```python
df["Genre"].value_counts()
# → Film     6
# → Série    4
# (trié par fréquence décroissante automatiquement)
```

### 🔎 `sort_values()` — trier les résultats

```python
df.sort_values("Note", ascending=False)  # du plus noté au moins noté
df.sort_values("Note", ascending=True)   # du moins noté au plus noté
df.sort_values("Année")                  # par défaut : croissant

# Trier sur plusieurs colonnes
df.sort_values(["Genre", "Note"], ascending=[True, False])
# → trie d'abord par Genre croissant, puis par Note décroissante
```

````python
import pandas as pd
df = pd.read_csv("netflix.csv")

# value_counts : combien de films vs séries ?
print("=== Répartition Films / Séries ===")
print(df["Genre"].value_counts())

print()

# Filtre simple : films uniquement
print("=== Films uniquement ===")
films = df[df["Genre"] == "Film"]
print(films[["Titre", "Année", "Note"]])

print()

# Filtre avec ET : séries avec une note >= 8
print("=== Séries bien notées (note >= 8) ===")
print(df[(df["Genre"] == "Série") & (df["Note"] >= 8)][["Titre", "Note"]])

print()

# Filtre avec OU : films OU sorties après 2019
print("=== Films OU sorties après 2019 ===")
print(df[(df["Genre"] == "Film") | (df["Année"] > 2019)][["Titre", "Genre", "Année"]])
````

````python
import pandas as pd
df = pd.read_csv("netflix.csv")

# Trier par note décroissante — classement général
print("=== Classement par note (meilleur en premier) ===")
print(df[["Titre", "Genre", "Note"]].sort_values("Note", ascending=False))

print()

# Trier par genre puis par note
print("=== Par genre, puis par note décroissante ===")
print(df[["Genre", "Titre", "Note"]].sort_values(
    ["Genre", "Note"],
    ascending=[True, False]   # Genre croissant, Note décroissante
))
````

## 📝 Partie 4 – Analyser avec `groupby()`

### 🔎 Rappel : `groupby()` = tableau croisé dynamique

On a vu `groupby()` dans le notebook 02. On l'applique ici sur les données Netflix.

```python
df.groupby("Genre")["Note"].mean()
# → "Pour chaque genre (Film / Série),
#    calcule la note moyenne"

df.groupby("Genre")[["Note", "Durée_min"]].mean()
# → Même chose sur deux colonnes

df.groupby("Langue")["Note"].agg(["mean", "count"])
# → Pour chaque langue : note moyenne + nombre d'œuvres
```

> 💡 **`count` dans `agg()`** compte le nombre de lignes dans chaque groupe — utile pour savoir si une moyenne est basée sur 1 ou sur 100 valeurs.

````python
import pandas as pd
df = pd.read_csv("netflix.csv")

# Comparaison Films vs Séries
print("=== Moyenne Note et Durée par genre ===")
print(df.groupby("Genre")[["Note", "Durée_min"]].mean().round(2))

print()

# Nombre d'œuvres + note moyenne par langue
print("=== Analyse par langue ===")
print(df.groupby("Langue")["Note"].agg(["mean", "count"]).round(2))

print()

# Analyse par année
print("=== Nombre de sorties et note moyenne par année ===")
print(df.groupby("Année")["Note"].agg(["count", "mean"]).round(2))

print()

# Meilleure note par genre
print("=== Meilleure note par genre ===")
print(df.groupby("Genre")["Note"].max())
````

## ✅ Récapitulatif

| Concept | Code | Ce qu'il faut retenir |
|---------|------|-----------------------|
| **`value_counts()`** | `df["col"].value_counts()` | Compte les occurrences de chaque valeur — trié par fréquence |
| **Filtre ET** | `df[(cond1) & (cond2)]` | Les deux conditions doivent être vraies — parenthèses obligatoires |
| **Filtre OU** | `df[(cond1) \| (cond2)]` | Au moins une condition vraie — parenthèses obligatoires |
| **`sort_values()`** | `df.sort_values("col", ascending=False)` | Trier — `False` = décroissant, `True` = croissant |
| **`groupby()`** | `df.groupby("col")["val"].mean()` | Regrouper et calculer par catégorie |
| **`agg()`** | `.agg(["mean", "count", "max"])` | Plusieurs fonctions d'un coup sur chaque groupe |

**Règle d'or des filtres combinés :**
- `&` = ET → les **deux** conditions doivent être vraies
- `|` = OU → **au moins une** condition doit être vraie
- Toujours mettre des **parenthèses** autour de chaque condition
- Jamais `and` / `or` avec des Series Pandas
