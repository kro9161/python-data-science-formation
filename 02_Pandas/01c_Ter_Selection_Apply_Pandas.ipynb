## 📝 Partie 1 – `loc[]` vs `iloc[]` : quelle différence ?

### 🔎 Le problème

On a déjà vu comment filtrer des lignes avec `df[condition]`. Mais parfois on veut être plus précis : accéder à **une ligne spécifique**, à **une cellule exacte**, ou sélectionner **des lignes ET des colonnes** en même temps. C'est là qu'interviennent `loc` et `iloc`.

| | `loc[]` | `iloc[]` |
|-|---------|----------|
| **Signification** | **L**ocation par **étiquette** (label) | **I**nteger **loc**ation — position numérique |
| **Lignes** | Par nom d'index (`0`, `"Alice"`) | Par numéro de position (`0`, `1`, `2`…) |
| **Colonnes** | Par **nom** de colonne (`"Nom"`, `"Salaire"`) | Par **numéro** de position (`0`, `1`, `2`…) |
| **⚠️ Fin de tranche** | **Incluse** : `loc[0:2]` → lignes 0, 1 **et 2** | **Exclue** : `iloc[0:2]` → lignes 0 **et 1** seulement |

**Le piège le plus fréquent — visualisation :**

```
     Nom    Salaire
0  Julie       3200    ← loc[0:2] : prend cette ligne ✓
1   Marc       5100    ← loc[0:2] : prend cette ligne ✓
2  Sonia       4100    ← loc[0:2] : prend cette ligne ✓ (FIN INCLUSE)
3  Rachid      4700    ← loc[0:2] : s'arrête AVANT
4   Nina       3500

loc[0:2]   → 3 lignes (0, 1, 2)  — fin incluse
iloc[0:2]  → 2 lignes (0, 1)     — fin exclue, comme Python
```

### 🔎 Syntaxe générale

```python
df.loc[lignes, colonnes]    # par étiquette (nom)
df.iloc[lignes, colonnes]   # par position numérique

# Exemples loc :
df.loc[2]                          # ligne d'index 2
df.loc[0:3, "Nom"]                 # lignes 0 à 3 inclus, colonne Nom
df.loc[0:3, ["Nom", "Salaire"]]    # lignes 0 à 3, 2 colonnes
df.loc[df["Service"] == "IT"]      # lignes où Service vaut IT

# Exemples iloc :
df.iloc[2]                         # 3e ligne (position 2)
df.iloc[0:3, 0:2]                  # lignes 0-1-2, colonnes 0 et 1
df.iloc[-1]                        # dernière ligne
df.iloc[-3:]                       # 3 dernières lignes
```

> 💡 **Règle simple :**
> - Vous connaissez les **noms** des colonnes → **`loc`**
> - Vous voulez les **n premières** ou **n dernières** lignes → **`iloc`**

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"     : ["Julie", "Marc", "Sonia", "Rachid", "Nina"],
    "Âge"     : [29, 45, 34, 38, 26],
    "Service" : ["RH", "IT", "Marketing", "IT", "Finance"],
    "Salaire" : [3200, 5100, 4100, 4700, 3500]
})

print("DataFrame de départ :")
print(employes)
print()

# loc — sélection par NOM
print("=== loc[2] : ligne d'index 2 ===")
print(employes.loc[2])

print()
print("=== loc[0:2, 'Nom'] : lignes 0 à 2 INCLUS, colonne Nom ===")
print(employes.loc[0:2, "Nom"])    # 3 lignes : 0, 1, 2

print()
print("=== loc[0:2, ['Nom','Salaire']] : lignes 0 à 2, 2 colonnes ===")
print(employes.loc[0:2, ["Nom", "Salaire"]])
````

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"     : ["Julie", "Marc", "Sonia", "Rachid", "Nina"],
    "Âge"     : [29, 45, 34, 38, 26],
    "Service" : ["RH", "IT", "Marketing", "IT", "Finance"],
    "Salaire" : [3200, 5100, 4100, 4700, 3500]
})

# iloc — sélection par POSITION
print("=== iloc[0] : 1ère ligne (position 0) ===")
print(employes.iloc[0])

print()
print("=== iloc[0:2] : positions 0 et 1 (FIN EXCLUE = 2 lignes) ===")
print(employes.iloc[0:2])

print()
print("=== Comparaison directe loc vs iloc sur [0:2] ===")
print(f"loc[0:2]  → {len(employes.loc[0:2])} lignes (fin incluse : 0, 1, 2)")
print(f"iloc[0:2] → {len(employes.iloc[0:2])} lignes (fin exclue  : 0, 1)")

print()
print("=== iloc[0:3, 0:2] : lignes 0-1-2, colonnes 0 et 1 ===")
print(employes.iloc[0:3, 0:2])

print()
print("=== iloc[-3:] : 3 dernières lignes ===")
print(employes.iloc[-3:])
````

## 📝 Partie 2 – `apply()` : transformer chaque valeur

### 🔎 Qu'est-ce que `apply()` ?

`apply()` permet d'appliquer une **fonction de votre choix** à chaque valeur d'une colonne. C'est comme faire une boucle `for` sur chaque cellule — mais en une seule ligne, sans écrire la boucle.

**Exemple concret :** catégoriser les salaires en Faible / Moyen / Élevé.

```python
# Sans apply — 8 lignes
niveaux = []
for salaire in df["Salaire"]:
    if salaire >= 5000:   niveaux.append("Élevé")
    elif salaire >= 4000: niveaux.append("Moyen")
    else:                 niveaux.append("Faible")
df["Niveau"] = niveaux

# Avec apply — 5 lignes, plus lisible
def niveau(s):
    if s >= 5000:   return "Élevé"
    elif s >= 4000: return "Moyen"
    else:           return "Faible"

df["Niveau"] = df["Salaire"].apply(niveau)
```

### 🔎 Les fonctions lambda — écrire une fonction en une ligne

Une **lambda** est une fonction anonyme compacte, parfaite pour les transformations simples :

```python
# Fonction classique
def doubler(x):
    return x * 2

# Équivalent lambda (même chose, une seule ligne)
lambda x: x * 2
#       ↑  ↑
#  argument  ce qu'on retourne

# Utilisation avec apply :
df["Salaire_x2"]  = df["Salaire"].apply(lambda x: x * 2)
df["Prix_TTC"]    = df["Prix_HT"].apply(lambda x: round(x * 1.20, 2))
df["Nom_majusc"]  = df["Nom"].apply(lambda x: x.upper())
```

### 🔎 `apply()` sur les lignes entières avec `axis=1`

Par défaut, `apply()` travaille sur **une colonne** (valeur par valeur).  
Avec `axis=1`, il travaille sur **chaque ligne** — on peut alors combiner plusieurs colonnes :

```python
# Besoin de 2 colonnes pour calculer → axis=1
df["Revenu_total"] = df.apply(
    lambda row: row["Salaire"] + row["Prime"],
    axis=1   # on passe chaque LIGNE à la fonction
)

# Équivalent avec une fonction classique :
def revenu_total(row):
    return row["Salaire"] + row["Prime"]

df["Revenu_total"] = df.apply(revenu_total, axis=1)
```

> 💡 **Règle simple :** besoin d'une seule colonne → `df["col"].apply(f)`. Besoin de plusieurs colonnes → `df.apply(f, axis=1)`.

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"     : ["Julie", "Marc", "Sonia", "Rachid", "Nina"],
    "Âge"     : [29, 45, 34, 38, 26],
    "Service" : ["RH", "IT", "Marketing", "IT", "Finance"],
    "Salaire" : [3200, 5100, 4100, 4700, 3500]
})

# apply() avec une fonction classique
def niveau_salaire(salaire):
    if salaire >= 5000:   return "Élevé"
    elif salaire >= 4000: return "Moyen"
    else:                 return "Faible"

employes["Niveau"] = employes["Salaire"].apply(niveau_salaire)

print("=== apply() avec une fonction classique ===")
print(employes)
````

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"     : ["Julie", "Marc", "Sonia", "Rachid", "Nina"],
    "Âge"     : [29, 45, 34, 38, 26],
    "Service" : ["RH", "IT", "Marketing", "IT", "Finance"],
    "Salaire" : [3200, 5100, 4100, 4700, 3500]
})

# apply() avec des lambdas — transformations rapides sur une colonne
employes["Prime (€)"]         = employes["Salaire"].apply(lambda x: round(x * 0.10, 2))
employes["Salaire_annuel"]    = employes["Salaire"].apply(lambda x: x * 12)
employes["Nom_majuscules"]    = employes["Nom"].apply(lambda x: x.upper())

print("=== apply() avec lambdas sur une colonne ===")
print(employes)

print()

# apply() avec axis=1 — combine plusieurs colonnes
employes["Revenu_total"] = employes.apply(
    lambda row: row["Salaire"] + row["Prime (€)"],
    axis=1
)

print("=== apply(axis=1) — combine Salaire + Prime ===")
print(employes[["Nom", "Salaire", "Prime (€)", "Revenu_total"]])
````

## ✅ Récapitulatif

| Concept | Syntaxe | Ce qu'il faut retenir |
|---------|---------|----------------------|
| **`loc[]`** | `df.loc[lignes, colonnes]` | Sélection par **nom** — fin de tranche **incluse** |
| **`iloc[]`** | `df.iloc[lignes, colonnes]` | Sélection par **position** — fin de tranche **exclue** |
| **`loc` + condition** | `df.loc[df["col"] > val, ["c1","c2"]]` | Filtrer les lignes ET choisir les colonnes en une fois |
| **`apply()` colonne** | `df["col"].apply(fonction)` | Applique une fonction à chaque valeur d'une colonne |
| **`apply()` lignes** | `df.apply(fonction, axis=1)` | Applique une fonction à chaque ligne (plusieurs colonnes disponibles) |
| **Lambda** | `lambda x: x * 2` | Fonction anonyme courte — équivalent d'une `def` simple en une ligne |

**La règle pour `axis=1` :** si votre fonction a besoin de lire **plusieurs colonnes** d'une même ligne → `axis=1`.
