## 📝 Partie 1 – Pourquoi visualiser ?

### 🔎 Un graphique dit en une seconde ce qu'un tableau dit en 5 minutes

Comparez :

```
Tableau des ventes par jour :          Courbe d'évolution :

Lundi    :  90                          240 ┤        ●
Mardi    : 110                          200 ┤     ●
Mercredi : 130                          160 ┤   ●        ●
Jeudi    : 150                          120 ┤ ●
Vendredi : 200                           80 ┤●
Samedi   : 240                               └──────────────
Dimanche : 180                               Lun Mar Mer Jeu Ven Sam Dim
```

Avec le tableau, on doit lire chaque chiffre. Avec la courbe, on voit **instantanément** que les ventes montent toute la semaine avec un pic le samedi.

### 🔎 `df.plot()` — la visualisation intégrée à Pandas

Pandas intègre directement un moteur de visualisation basé sur Matplotlib. On n'a pas besoin de savoir utiliser Matplotlib — on appelle simplement `.plot()` sur un DataFrame ou une Series.

```python
df["col"].plot()                     # graphique par défaut (courbe)
df["col"].plot(kind="bar")           # graphique en barres
df.plot(kind="scatter", x="a", y="b") # nuage de points
```

### 🔎 Quel type de graphique pour quelle question ?

| Question | Type | `kind=` |
|----------|------|---------|
| Comment une valeur **évolue dans le temps** ? | Courbe | `"line"` |
| **Comparer** des catégories entre elles ? | Barres | `"bar"` ou `"barh"` |
| Quelle est la **répartition en %** d'une catégorie ? | Camembert | `"pie"` |
| Comment se **distribuent** les valeurs ? | Histogramme | `"hist"` |
| Y a-t-il une **relation** entre deux variables ? | Nuage de points | `"scatter"` |

**Équivalents Excel :**
- `"line"` = Courbes
- `"bar"` = Histogramme (colonnes)
- `"pie"` = Secteurs (camembert)
- `"hist"` = Histogramme (distribution)
- `"scatter"` = Nuage de points

````python
import pandas as pd

# Jeu de données : ventes d'une pizzeria sur 7 jours
pizza = pd.DataFrame({
    "Jour"           : ["Lundi", "Mardi", "Mercredi", "Jeudi",
                        "Vendredi", "Samedi", "Dimanche"],
    "Pizzas_vendues" : [90, 110, 130, 150, 200, 240, 180],
    "Prix_moyen"     : [10.5, 11.0, 11.2, 11.4, 11.7, 12.0, 11.9],
    "Type_jour"      : ["Semaine", "Semaine", "Semaine", "Semaine",
                        "Week-end", "Week-end", "Week-end"]
})

# Jeu de données : catalogue Netflix
netflix = pd.DataFrame({
    "Titre"    : ["Inception", "The Crown", "Stranger Things", "Roma",
                  "Extraction", "The Irishman", "Emily in Paris",
                  "The Witcher", "Bird Box", "Enola Holmes"],
    "Genre"    : ["Film", "Série", "Série", "Film", "Film",
                  "Film", "Série", "Série", "Film", "Film"],
    "Année"    : [2010, 2016, 2017, 2018, 2020, 2019, 2020, 2019, 2018, 2020],
    "Durée_min": [148, 50, 55, 135, 117, 209, 30, 60, 124, 123],
    "Note"     : [8.8, 8.6, 8.7, 7.9, 6.8, 7.8, 6.5, 8.2, 6.6, 6.9]
})

print("Données pizza :")
print(pizza)
print()
print("Données Netflix :")
print(netflix.head())
````

## 📝 Partie 2 – La courbe : évolution dans le temps

### 🔎 Quand l'utiliser ?

La courbe est parfaite pour montrer **comment une valeur change dans le temps** — ventes par semaine, température par mois, cours boursier par jour.

**Question type :** *"Comment évoluent les ventes au fil de la semaine ?"*

### 🔎 Options principales

```python
series.plot(
    kind    = "line",          # type de graphique (défaut si omis)
    title   = "Mon titre",     # titre du graphique
    xlabel  = "Jours",         # étiquette axe X
    ylabel  = "Pizzas",        # étiquette axe Y
    color   = "steelblue",     # couleur de la courbe
    marker  = "o",             # point visible à chaque valeur
    figsize = (10, 4)          # taille du graphique (largeur, hauteur) en pouces
)
```

````python
import pandas as pd
import matplotlib.pyplot as plt

pizza = pd.DataFrame({
    "Jour"           : ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"],
    "Pizzas_vendues" : [90, 110, 130, 150, 200, 240, 180],
    "Prix_moyen"     : [10.5, 11.0, 11.2, 11.4, 11.7, 12.0, 11.9]
})

# Courbe simple : évolution des ventes
pizza.plot(
    x       = "Jour",
    y       = "Pizzas_vendues",
    kind    = "line",
    title   = "Ventes de pizzas dans la semaine",
    xlabel  = "Jour",
    ylabel  = "Pizzas vendues",
    color   = "steelblue",
    marker  = "o",
    figsize = (9, 4)
)
plt.tight_layout()
plt.show()

# Deux courbes sur le même graphique
pizza.plot(
    x       = "Jour",
    y       = ["Pizzas_vendues", "Prix_moyen"],
    kind    = "line",
    title   = "Ventes et Prix sur la semaine",
    marker  = "o",
    figsize = (9, 4)
)
plt.tight_layout()
plt.show()
````

## 📝 Partie 3 – Le graphique en barres : comparer des catégories

### 🔎 Quand l'utiliser ?

Les barres sont parfaites pour **comparer des catégories** entre elles — combien de films vs séries, quel produit se vend le mieux, quel service a le plus d'employés.

**Question type :** *"Quel genre (Film ou Série) est le plus représenté ?"*

```python
kind = "bar"   # barres verticales
kind = "barh"  # barres horizontales (utile si les étiquettes sont longues)
```

````python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.DataFrame({
    "Genre"    : ["Film", "Série", "Série", "Film", "Film",
                  "Film", "Série", "Série", "Film", "Film"],
    "Année"    : [2010, 2016, 2017, 2018, 2020, 2019, 2020, 2019, 2018, 2020],
    "Durée_min": [148, 50, 55, 135, 117, 209, 30, 60, 124, 123],
    "Note"     : [8.8, 8.6, 8.7, 7.9, 6.8, 7.8, 6.5, 8.2, 6.6, 6.9]
})

# Barres verticales : nombre de films et séries
netflix["Genre"].value_counts().plot(
    kind    = "bar",
    title   = "Nombre de films et séries dans le catalogue",
    xlabel  = "Genre",
    ylabel  = "Nombre d'œuvres",
    color   = ["steelblue", "orange"],
    figsize = (7, 4)
)
plt.tight_layout()
plt.show()

# Barres horizontales : durée moyenne par genre (plus lisible avec des étiquettes)
netflix.groupby("Genre")["Durée_min"].mean().plot(
    kind    = "barh",
    title   = "Durée moyenne par genre (minutes)",
    xlabel  = "Minutes",
    color   = "coral",
    figsize = (7, 3)
)
plt.tight_layout()
plt.show()
````

## 📝 Partie 4 – Le camembert : les proportions

### 🔎 Quand l'utiliser ?

Le camembert montre la **répartition en pourcentage** d'un ensemble. Il est utile quand on veut voir quelle part du total représente chaque catégorie.

**Question type :** *"Quelle proportion du catalogue est des films vs des séries ?"*

> ⚠️ **Limite :** le camembert devient difficile à lire avec plus de 5-6 catégories. Préférez les barres dans ce cas.

```python
series.plot(
    kind     = "pie",
    autopct  = "%1.1f%%",  # affiche les pourcentages sur chaque part
    ylabel   = "",         # supprime l'étiquette verticale (parasite par défaut)
    startangle = 90        # commence à 12h au lieu de 3h
)
```

````python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.DataFrame({
    "Genre" : ["Film", "Série", "Série", "Film", "Film",
               "Film", "Série", "Série", "Film", "Film"]
})

# Camembert : répartition Films / Séries
netflix["Genre"].value_counts().plot(
    kind       = "pie",
    autopct    = "%1.1f%%",   # affiche ex: 60.0%
    ylabel     = "",           # supprime l'étiquette verticale inutile
    startangle = 90,
    colors     = ["steelblue", "orange"],
    figsize    = (5, 5),
    title      = "Répartition Films / Séries"
)
plt.tight_layout()
plt.show()
````

## 📝 Partie 5 – L'histogramme : la distribution

### 🔎 Qu'est-ce qu'un histogramme ?

L'histogramme répond à la question : **"Comment les valeurs se répartissent-elles ?"**

Il découpe les valeurs en **intervalles** (appelés "bins") et compte combien de valeurs tombent dans chaque intervalle.

**Exemple :** pour 10 notes entre 6.5 et 8.8, avec `bins=4` :

```
Intervalle   Notes qui tombent dedans      Hauteur barre
[6.5 – 7.1]  6.5, 6.6, 6.8, 6.9           4 notes
[7.1 – 7.7]  (aucune)                      0 notes
[7.7 – 8.3]  7.8, 7.9, 8.2                3 notes
[8.3 – 8.9]  8.6, 8.7, 8.8                3 notes
```

**Question type :** *"Les notes sont-elles concentrées ou dispersées ?"*

> 💡 **`bins`** contrôle le nombre d'intervalles. Trop peu de bins → on perd du détail. Trop de bins → le graphique devient illisible. 8-15 bins est souvent un bon départ.

**Équivalent Excel :** Données → Analyse → Histogramme.

````python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.DataFrame({
    "Durée_min": [148, 50, 55, 135, 117, 209, 30, 60, 124, 123],
    "Note"     : [8.8, 8.6, 8.7, 7.9, 6.8, 7.8, 6.5, 8.2, 6.6, 6.9]
})

# Histogramme des notes
netflix["Note"].plot(
    kind    = "hist",
    bins    = 5,
    title   = "Distribution des notes Netflix",
    xlabel  = "Note",
    ylabel  = "Nombre d'œuvres",
    color   = "steelblue",
    edgecolor = "white",    # séparation visible entre les barres
    figsize = (7, 4)
)
plt.tight_layout()
plt.show()

# Histogramme des durées
netflix["Durée_min"].plot(
    kind    = "hist",
    bins    = 5,
    title   = "Distribution des durées",
    xlabel  = "Durée (minutes)",
    ylabel  = "Nombre d'œuvres",
    color   = "coral",
    edgecolor = "white",
    figsize = (7, 4)
)
plt.tight_layout()
plt.show()
````

## 📝 Partie 6 – Le nuage de points : les relations

### 🔎 Quand l'utiliser ?

Le nuage de points permet de voir s'il existe une **relation entre deux variables** numériques. Chaque point représente une observation.

**Question type :** *"Les films longs sont-ils mieux notés que les films courts ?"*

```
Si les points forment une ligne montante → corrélation positive
  (quand X augmente, Y augmente aussi)

Si les points forment une ligne descendante → corrélation négative
  (quand X augmente, Y diminue)

Si les points sont éparpillés → pas de relation claire
```

**C'est le lien avec la corrélation qu'on a vue dans NumPy !** Le nuage de points en est la visualisation.

````python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.DataFrame({
    "Titre"    : ["Inception", "The Crown", "Stranger Things", "Roma",
                  "Extraction", "The Irishman", "Emily in Paris",
                  "The Witcher", "Bird Box", "Enola Holmes"],
    "Durée_min": [148, 50, 55, 135, 117, 209, 30, 60, 124, 123],
    "Note"     : [8.8, 8.6, 8.7, 7.9, 6.8, 7.8, 6.5, 8.2, 6.6, 6.9]
})

# Nuage de points : Durée vs Note
netflix.plot(
    kind    = "scatter",
    x       = "Durée_min",
    y       = "Note",
    title   = "Relation entre durée et note",
    xlabel  = "Durée (minutes)",
    ylabel  = "Note",
    color   = "steelblue",
    s       = 80,           # taille des points
    figsize = (7, 5)
)
plt.tight_layout()
plt.show()

# Calcul de la corrélation pour confirmer
r = netflix["Durée_min"].corr(netflix["Note"])
print(f"Corrélation Durée ↔ Note : {r:.3f}")
print("→ Valeur proche de 0 : pas de relation claire entre durée et note")
````

````python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.DataFrame({
    "Genre"    : ["Film", "Série", "Série", "Film", "Film",
                  "Film", "Série", "Série", "Film", "Film"],
    "Année"    : [2010, 2016, 2017, 2018, 2020, 2019, 2020, 2019, 2018, 2020],
    "Durée_min": [148, 50, 55, 135, 117, 209, 30, 60, 124, 123],
    "Note"     : [8.8, 8.6, 8.7, 7.9, 6.8, 7.8, 6.5, 8.2, 6.6, 6.9]
})

# TODO 1 : barres note moyenne par genre
netflix.groupby(...)[...].mean().plot(
    kind=..., title="Note moyenne par genre",
    ylabel="Note", color=["steelblue", "orange"], figsize=(6, 4)
)
plt.tight_layout(); plt.show()

# TODO 2 : histogramme durées
netflix[...].plot(
    kind=..., bins=..., title="Distribution des durées",
    xlabel="Durée (min)", edgecolor="white", figsize=(7, 4)
)
plt.tight_layout(); plt.show()

# TODO 3 : courbe note moyenne par année
netflix.groupby(...)[...].mean().plot(
    kind=..., marker=..., title="Note moyenne par année", figsize=(7, 4)
)
plt.tight_layout(); plt.show()

# TODO 4 : camembert genres
netflix[...].value_counts().plot(
    kind=..., autopct=..., ylabel=..., figsize=(5, 5)
)
plt.tight_layout(); plt.show()

# TODO 5 : BONUS — grille 2×2
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

netflix.groupby("Genre")["Note"].mean().plot(kind="bar",   ax=axes[0,0], title="Note par genre",  color=["steelblue","orange"])
netflix["Durée_min"].plot(            kind="hist",  ax=axes[0,1], title="Distribution durées", bins=6, edgecolor="white")
netflix.groupby("Année")["Note"].mean().plot(kind="line",  ax=axes[1,0], title="Note par année",  marker="o")
netflix["Genre"].value_counts().plot( kind="pie",   ax=axes[1,1], title="Répartition genres",  autopct="%1.1f%%", ylabel="")

plt.suptitle("Tableau de bord Netflix", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
````

## ✅ Récapitulatif

### Quel graphique pour quelle question ?
