## 📝 Partie 1 – Le DataFrame : c'est quoi ?

### 🔎 Series vs DataFrame

Dans le notebook précédent, on a vu la **Series** : une seule colonne de données.

Un **DataFrame**, c'est **plusieurs Series collées côte à côte** — exactement comme un tableau Excel avec des lignes et des colonnes.

```
Series "Nom" :      Series "Âge" :      Series "Ville" :
  0 → Alice           0 → 25              0 → Paris
  1 → Bob             1 → 32              1 → Lyon
  2 → Chloé           2 → 29              2 → Marseille

DataFrame (les 3 Series ensemble) :

     Nom   Âge       Ville
0  Alice    25       Paris
1    Bob    32        Lyon
2  Chloé    29   Marseille
↑
index (partagé par toutes les colonnes)
```

### 🔎 Pourquoi un dictionnaire pour créer un DataFrame ?

La façon la plus courante de créer un DataFrame "à la main" est d'utiliser un **dictionnaire Python** :
- chaque **clé** du dictionnaire = un **nom de colonne**
- chaque **valeur** (liste) = les **données de cette colonne**

```python
data = {
    "Nom" : ["Alice", "Bob", "Chloé"],   # colonne 1 : 3 noms
    "Âge" : [25, 32, 29],                # colonne 2 : 3 âges
    "Ville": ["Paris", "Lyon", "Nice"]   # colonne 3 : 3 villes
}
# → toutes les listes doivent avoir la MÊME longueur

df = pd.DataFrame(data)  # on convertit le dictionnaire en DataFrame
```

> ⚠️ **Règle importante :** toutes les listes doivent avoir exactement le même nombre d'éléments. Si une colonne a 3 valeurs et une autre en a 4, Pandas génère une erreur.

````python
import pandas as pd

# On crée un DataFrame représentant des employés
employes = pd.DataFrame({
    "Nom"    : ["Alice", "Bob", "Chloé", "David", "Eva"],
    "Âge"    : [25, 32, 29, 41, 36],           # entiers
    "Service": ["IT", "RH", "IT", "Marketing", "RH"],
    "Salaire": [3200, 2900, 4100, 4700, 3500]   # en euros
})

# Afficher le DataFrame
print("Notre DataFrame :")
employes
````

## 📝 Partie 2 – Explorer un DataFrame

### 🔎 Les commandes d'exploration essentielles

Quand on reçoit un nouveau jeu de données (fichier CSV, base de données…), les **premières choses à faire** sont toujours les mêmes :

| Méthode | Ce qu'elle fait | Équivalent Excel |
|---------|----------------|------------------|
| `df.head(n)` | Affiche les **n premières lignes** (5 par défaut) | Faire défiler le début du fichier |
| `df.tail(n)` | Affiche les **n dernières lignes** | Faire défiler la fin |
| `df.shape` | Retourne **(nb_lignes, nb_colonnes)** | Voir le nombre de lignes/colonnes |
| `df.info()` | Affiche le **type de chaque colonne** et les valeurs manquantes | Vérifier les types de données |
| `df.describe()` | **Statistiques** sur les colonnes numériques | Statistiques descriptives |
| `df.columns` | Liste les **noms des colonnes** | Voir les en-têtes de colonnes |
| `df.dtypes` | Type de données de chaque colonne | — |

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"    : ["Alice", "Bob", "Chloé", "David", "Eva"],
    "Âge"    : [25, 32, 29, 41, 36],
    "Service": ["IT", "RH", "IT", "Marketing", "RH"],
    "Salaire": [3200, 2900, 4100, 4700, 3500]
})

# head() : les premières lignes (aperçu du tableau)
print("=== head(3) : 3 premières lignes ===")
print(employes.head(3))

print()

# shape : dimensions du tableau
print("=== shape ===")
print(employes.shape)              # (5, 4) → 5 lignes, 4 colonnes
print("Lignes   :", employes.shape[0])
print("Colonnes :", employes.shape[1])

print()

# columns : noms des colonnes
print("=== columns ===")
print(employes.columns.tolist())
````

````python
# info() : la "carte d'identité" du DataFrame
# Indique le type de chaque colonne et s'il manque des valeurs
employes.info()
````

### 🔎 Lire le résultat de `info()`

```
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 5 entries, 0 to 4     ← 5 lignes (index de 0 à 4)
Data columns (total 4 columns):   ← 4 colonnes
 #   Column    Non-Null Count  Dtype
---  ------    --------------  -----
 0   Nom       5 non-null      object    ← texte, 0 valeur manquante
 1   Âge       5 non-null      int64     ← entier, 0 valeur manquante
 2   Service   5 non-null      object    ← texte
 3   Salaire   5 non-null      int64     ← entier
```

**Les types de colonnes :**
- `object` → texte (chaîne de caractères)
- `int64` → nombre entier
- `float64` → nombre décimal
- `bool` → True / False
- `datetime64` → date et heure

> 💡 **Si `Non-Null Count` est inférieur au nombre total de lignes**, ça veut dire qu'il y a des **valeurs manquantes** dans cette colonne — on verra comment les gérer dans le notebook `01_Bis`.

````python
# describe() : statistiques sur les colonnes numériques
# Ici : Âge et Salaire
employes.describe()
````

## 📝 Partie 3 – Sélectionner des colonnes

### 🔎 Une colonne → `df["nom_colonne"]`

Pour récupérer **une seule colonne**, on utilise les crochets avec le nom de la colonne entre guillemets. Le résultat est une **Series** (une colonne).

```python
df["Nom"]      # → Series contenant les noms
df["Salaire"]  # → Series contenant les salaires
```

### 🔎 Plusieurs colonnes → `df[["col1", "col2"]]`

Pour récupérer **plusieurs colonnes**, on passe une **liste** de noms. Le résultat est un **DataFrame** (un sous-tableau).

```python
df[["Nom", "Salaire"]]   # → DataFrame avec 2 colonnes seulement
```

> 💡 **Attention aux doubles crochets !**  
> - `df["Nom"]` → 1 paire de crochets → retourne une **Series**  
> - `df[["Nom", "Salaire"]]` → 2 paires de crochets → retourne un **DataFrame**

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"    : ["Alice", "Bob", "Chloé", "David", "Eva"],
    "Âge"    : [25, 32, 29, 41, 36],
    "Service": ["IT", "RH", "IT", "Marketing", "RH"],
    "Salaire": [3200, 2900, 4100, 4700, 3500]
})

# Une seule colonne → Series
print("=== Une colonne (Series) ===")
print(employes["Nom"])
print("Type :", type(employes["Nom"]))

print()

# Plusieurs colonnes → DataFrame
print("=== Deux colonnes (DataFrame) ===")
print(employes[["Nom", "Salaire"]])
print("Type :", type(employes[["Nom", "Salaire"]]))

print()

# Calcul direct sur une colonne
print("=== Calculs sur la colonne Salaire ===")
print(f"Salaire moyen  : {employes['Salaire'].mean():.0f} €")
print(f"Salaire maxi   : {employes['Salaire'].max()} €")
print(f"Salaire mini   : {employes['Salaire'].min()} €")
````

## 📝 Partie 4 – Ajouter et supprimer des colonnes

### 🔎 Ajouter une colonne

Pour ajouter une nouvelle colonne, on l'assigne directement comme si elle existait déjà :

```python
# Ajouter une colonne calculée (salaire annuel)
df["Salaire_annuel"] = df["Salaire"] * 12

# Ajouter une colonne avec des valeurs fixes
df["Statut"] = "Actif"
```

> 💡 C'est exactement comme ajouter une colonne dans Excel avec une formule : `=B2*12`

### 🔎 Supprimer une colonne — `drop()`

```python
# Supprimer une colonne
df = df.drop("Statut", axis=1)
#                        ↑
#               axis=1 = on travaille sur les colonnes
#               axis=0 = on travaillerait sur les lignes
```

**Pourquoi `axis=1` ?** Pandas utilise deux axes :
- `axis=0` → direction verticale (les **lignes**)
- `axis=1` → direction horizontale (les **colonnes**)

```
       col0   col1   col2
          ↕    ↕     ↕
ligne0  [  A    B     C  ]  ← axis=0 : on supprime une ligne entière
ligne1  [  D    E     F  ]  ← axis=1 : on supprime une colonne entière
        ←────────────────→
              axis=1
```

> ⚠️ **Important :** `drop()` ne modifie pas le DataFrame original — il retourne une **copie** avec la suppression. Pour que la suppression soit permanente, il faut réassigner : `df = df.drop(...)`

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"    : ["Alice", "Bob", "Chloé", "David", "Eva"],
    "Âge"    : [25, 32, 29, 41, 36],
    "Service": ["IT", "RH", "IT", "Marketing", "RH"],
    "Salaire": [3200, 2900, 4100, 4700, 3500]
})

print("=== DataFrame original ===")
print(employes)

# Ajouter une colonne calculée : salaire annuel
employes["Salaire_annuel"] = employes["Salaire"] * 12
print("\n=== Après ajout de Salaire_annuel ===")
print(employes)

# Ajouter une colonne avec valeur fixe
employes["Statut"] = "CDI"
print("\n=== Après ajout de Statut ===")
print(employes)

# Supprimer la colonne Statut
employes = employes.drop("Statut", axis=1)   # axis=1 = colonne
print("\n=== Après suppression de Statut ===")
print(employes)
````

## 📝 Partie 5 – Filtrer les lignes

### 🔎 Sélectionner des lignes selon une condition

C'est l'opération la plus courante en analyse de données — l'équivalent du **filtre automatique** dans Excel.

La syntaxe : `df[condition]`

```python
# Employés du service IT
df[df["Service"] == "IT"]

# Employés avec un salaire > 4000 €
df[df["Salaire"] > 4000]

# Employés du service IT ET avec un salaire > 4000 €
df[(df["Service"] == "IT") & (df["Salaire"] > 4000)]
#   ↑ parenthèses obligatoires autour de chaque condition
```

> 💡 **Comment ça marche :**  
> `df["Service"] == "IT"` crée une Series de True/False (un masque).  
> Quand on la passe à `df[...]`, Pandas ne garde que les lignes où c'est True.
> C'est exactement le même principe que le filtrage NumPy qu'on a vu précédemment.

````python
import pandas as pd

employes = pd.DataFrame({
    "Nom"    : ["Alice", "Bob", "Chloé", "David", "Eva"],
    "Âge"    : [25, 32, 29, 41, 36],
    "Service": ["IT", "RH", "IT", "Marketing", "RH"],
    "Salaire": [3200, 2900, 4100, 4700, 3500]
})

# Filtrer : employés du service IT
print("=== Employés IT ===")
print(employes[employes["Service"] == "IT"])

print()

# Filtrer : salaire > 3500 €
print("=== Salaire > 3500 € ===")
print(employes[employes["Salaire"] > 3500])

print()

# Double condition : IT ET salaire > 3000 €
print("=== IT avec salaire > 3000 € ===")
masque = (employes["Service"] == "IT") & (employes["Salaire"] > 3000)
print(employes[masque])

print()

# Combien d'employés correspondent au filtre ?
print(f"Nombre d'employés IT avec salaire > 3000 € : {masque.sum()}")
````

## ✅ Récapitulatif

| Concept | Ce qu'il faut retenir |
|---------|----------------------|
| **DataFrame** | Tableau avec lignes et colonnes — équivalent d'une feuille Excel |
| **Créer** | `pd.DataFrame({"col1": [...], "col2": [...]})` — toutes les listes doivent avoir la même longueur |
| **`head(n)` / `tail(n)`** | Aperçu des n premières / dernières lignes |
| **`shape`** | `(nb_lignes, nb_colonnes)` — attribut sans parenthèses |
| **`info()`** | Type de chaque colonne + valeurs manquantes |
| **`describe()`** | Statistiques sur les colonnes numériques |
| **`df["col"]`** | Sélectionner une colonne → retourne une **Series** |
| **`df[["c1", "c2"]]`** | Sélectionner plusieurs colonnes → retourne un **DataFrame** |
| **`df["nouvelle"] = ...`** | Ajouter une colonne calculée |
| **`df.drop("col", axis=1)`** | Supprimer une colonne (`axis=1` = colonne, `axis=0` = ligne) |
| **`df[condition]`** | Filtrer les lignes selon une condition — équivalent filtre Excel |
| **`idxmax()` / `idxmin()`** | Retourne l'indice de la valeur max / min |
