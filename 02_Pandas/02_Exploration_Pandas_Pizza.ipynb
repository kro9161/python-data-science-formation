## 📝 Partie 1 – Le fichier CSV : c'est quoi ?

### 🔎 CSV — Comma Separated Values

Un **fichier CSV** est un fichier texte simple où les données sont organisées en lignes, et les colonnes sont séparées par des virgules (ou des points-virgules en France).

**Exemple — ce qu'on voit dans Excel :**

```
| Date       | Jour     | Pizzas | Prix_moyen | Pluie |
| 2024-03-01 | Vendredi |    120 |       11.5 | False |
| 2024-03-02 | Samedi   |    200 |       12.0 | False |
```

**Ce que contient vraiment le fichier .csv :**

```
Date,Jour,Pizzas_vendues,Prix_moyen,Pluie
2024-03-01,Vendredi,120,11.5,False
2024-03-02,Samedi,200,12.0,False
2024-03-03,Dimanche,240,12.5,False
```

C'est le format d'échange universel entre logiciels : Excel, Python, R, SQL, Google Sheets — tous lisent et écrivent des CSV.

### 🔎 Lire un CSV avec Pandas

```python
df = pd.read_csv("nom_du_fichier.csv")   # le fichier doit être dans le même dossier
```

En une ligne, Pandas lit le fichier, détecte automatiquement les colonnes, devine les types de données, et crée un DataFrame prêt à l'emploi.

**Équivalent Excel :** Fichier → Ouvrir → Sélectionner le CSV → Importer.

> 💡 **Important :** le fichier CSV doit se trouver dans le **même dossier** que votre notebook, ou vous devez indiquer le chemin complet.

````python
import pandas as pd

# On crée d'abord le jeu de données pizza et on le sauvegarde en CSV
# (dans un vrai projet, ce fichier viendrait d'un export de caisse, d'une BDD, etc.)

ventes_pizza = pd.DataFrame({
    "Date"           : ["2024-03-01", "2024-03-02", "2024-03-03",
                        "2024-03-04", "2024-03-05", "2024-03-06", "2024-03-07"],
    "Jour"           : ["Vendredi", "Samedi", "Dimanche", "Lundi",
                        "Mardi", "Mercredi", "Jeudi"],
    "Type_jour"      : ["Week-end", "Week-end", "Week-end",
                        "Semaine", "Semaine", "Semaine", "Semaine"],
    "Pizzas_vendues" : [120, 200, 240, 90, 110, 95, 130],
    "Prix_moyen"     : [11.5, 12.0, 12.5, 10.5, 11.0, 10.8, 11.2],
    "Pluie"          : [False, False, False, True, True, False, False]
})

# Sauvegarder dans le dossier courant (même endroit que le notebook)
ventes_pizza.to_csv("ventes_pizza.csv", index=False)
print("✅ Fichier 'ventes_pizza.csv' créé dans le dossier courant.")
print(f"   {len(ventes_pizza)} lignes, {len(ventes_pizza.columns)} colonnes")
````

## 📝 Partie 2 – Lire et explorer le CSV

### 🔎 Les premières choses à faire quand on reçoit un fichier

En data science, quand on reçoit un nouveau fichier de données, on suit toujours le même rituel d'exploration :

```
1. Lire le fichier          → pd.read_csv()
2. Aperçu des premières lignes → head()
3. Structure et types          → info()
4. Statistiques rapides        → describe()
5. Valeurs manquantes          → isna().sum()
```

Ces 5 étapes donnent une vision complète du fichier en moins d'une minute.

````python
import pandas as pd

# Lire le CSV — Pandas détecte automatiquement les types
df = pd.read_csv("ventes_pizza.csv")

# 1. Aperçu des premières lignes
print("=== head() : aperçu du tableau ===")
print(df.head())

print()

# 2. Dimensions
print(f"=== shape : {df.shape[0]} lignes × {df.shape[1]} colonnes ===")
````

````python
import pandas as pd
df = pd.read_csv("ventes_pizza.csv")

# 3. Structure et types de colonnes
print("=== info() : structure du tableau ===")
df.info()

print()

# 4. Valeurs manquantes
print("=== Valeurs manquantes par colonne ===")
print(df.isna().sum())
````

````python
import pandas as pd
df = pd.read_csv("ventes_pizza.csv")

# 5. Statistiques sur les colonnes numériques
print("=== describe() : statistiques ===")
print(df.describe())

print()
print("Lecture rapide des résultats :")
print(f"  → Pizzas vendues en moyenne : {df['Pizzas_vendues'].mean():.0f} / jour")
print(f"  → Meilleur jour             : {df['Pizzas_vendues'].max()} pizzas")
print(f"  → Pire jour                 : {df['Pizzas_vendues'].min()} pizzas")
print(f"  → Prix moyen                : {df['Prix_moyen'].mean():.2f} €")
````

## 📝 Partie 3 – Filtrer, trier, calculer

### 🔎 Rappel des opérations de base

On applique ici les notions vues dans les notebooks précédents sur des données réelles :

| Question | Code Pandas | Équivalent Excel |
|----------|------------|------------------|
| Jours pluvieux ? | `df[df["Pluie"] == True]` | Filtre sur colonne Pluie |
| Meilleur jour ? | `df.loc[df["Pizzas_vendues"].idxmax(), "Jour"]` | `=INDEX(EQUIV(MAX(...)))` |
| Trier par ventes | `df.sort_values("Pizzas_vendues", ascending=False)` | Trier Z→A |
| Ajouter CA | `df["CA"] = df["Pizzas_vendues"] * df["Prix_moyen"]` | `=B2*C2` glissé vers le bas |

````python
import pandas as pd
df = pd.read_csv("ventes_pizza.csv")

# Filtrer : jours de pluie
print("=== Jours de pluie ===")
print(df[df["Pluie"] == True])

print()

# Filtrer : jours où on a vendu plus de 150 pizzas
print("=== Jours > 150 pizzas ===")
print(df[df["Pizzas_vendues"] > 150])

print()

# Quel jour a-t-on vendu le plus ?
idx_max = df["Pizzas_vendues"].idxmax()
meilleur_jour = df.loc[idx_max, "Jour"]
record = df.loc[idx_max, "Pizzas_vendues"]
print(f"=== Meilleur jour : {meilleur_jour} avec {record} pizzas ===")
````

````python
import pandas as pd
df = pd.read_csv("ventes_pizza.csv")

# Trier par ventes décroissantes
print("=== Classement des jours par ventes ===")
print(df[["Jour", "Pizzas_vendues"]].sort_values("Pizzas_vendues", ascending=False))

print()

# Ajouter une colonne calculée : chiffre d'affaires
# Équivalent Excel : colonne F avec formule =D2*E2
df["CA (€)"] = df["Pizzas_vendues"] * df["Prix_moyen"]
print("=== Tableau avec Chiffre d'Affaires ===")
print(df[["Jour", "Pizzas_vendues", "Prix_moyen", "CA (€)"]].round(2))

print()
print(f"CA total sur la semaine : {df['CA (€)'].sum():.2f} €")
print(f"CA moyen par jour       : {df['CA (€)'].mean():.2f} €")
````

## 📝 Partie 4 – `groupby()` : regrouper les données

### 🔎 Qu'est-ce que `groupby()` ?

`groupby()` est l'une des fonctions les plus puissantes de Pandas. Elle permet de **regrouper les lignes** selon une colonne, puis d'**appliquer un calcul** sur chaque groupe.

C'est exactement l'équivalent d'un **tableau croisé dynamique** dans Excel.

**Exemple concret :** on veut comparer les ventes en semaine vs le week-end.

```
Données brutes :                   Après groupby("Type_jour").mean() :

Type_jour   Pizzas_vendues         Type_jour   Pizzas_vendues
Semaine          90                Semaine           106
Semaine         110                Week-end          187
Semaine          95
Semaine         130
Week-end        120
Week-end        200
Week-end        240
```

### 🔎 Syntaxe

```python
df.groupby("colonne_de_groupe")["colonne_à_calculer"].fonction()

# Exemples :
df.groupby("Type_jour")["Pizzas_vendues"].mean()   # moyenne par type de jour
df.groupby("Type_jour")["CA"].sum()                # CA total par type de jour
df.groupby("Type_jour")["Pizzas_vendues"].max()    # record par type de jour

# Plusieurs colonnes à calculer :
df.groupby("Type_jour")[["Pizzas_vendues", "CA"]].mean()
```

> 💡 **Traduction en langage humain :** `groupby("Type_jour")["Pizzas_vendues"].mean()` = "pour chaque type de jour, calcule la moyenne des pizzas vendues".

````python
import pandas as pd
df = pd.read_csv("ventes_pizza.csv")
df["CA (€)"] = df["Pizzas_vendues"] * df["Prix_moyen"]

# groupby : moyenne des ventes par type de jour
print("=== Moyenne des pizzas vendues : semaine vs week-end ===")
print(df.groupby("Type_jour")["Pizzas_vendues"].mean().round(0))

print()

# groupby : plusieurs statistiques
print("=== Statistiques complètes par type de jour ===")
print(df.groupby("Type_jour")[["Pizzas_vendues", "CA (€)"]].agg(["mean", "sum", "max"]).round(2))

print()

# groupby : impact de la pluie sur les ventes
print("=== Ventes moyennes : avec pluie vs sans pluie ===")
print(df.groupby("Pluie")["Pizzas_vendues"].mean().round(0))
````

## ✅ Récapitulatif

| Concept | Code | Ce qu'il faut retenir |
|---------|------|-----------------------|
| **CSV** | `pd.read_csv("fichier.csv")` | Fichier texte universel — Pandas le lit en une ligne |
| **Exploration** | `head()`, `info()`, `describe()`, `isna().sum()` | Rituel à faire systématiquement sur tout nouveau fichier |
| **Colonne calculée** | `df["CA"] = df["col1"] * df["col2"]` | Équivalent d'une formule Excel glissée sur toute la colonne |
| **Tri** | `sort_values("col", ascending=False)` | Trier par ordre décroissant — `ascending=True` pour croissant |
| **`groupby()`** | `df.groupby("col")["val"].mean()` | Regrouper et calculer — équivalent du tableau croisé dynamique Excel |
| **`agg()`** | `.agg(["mean", "sum", "max"])` | Appliquer plusieurs fonctions d'un coup sur un groupe |
| **Sauvegarder** | `df.to_csv("fichier.csv", index=False)` | Exporter le DataFrame en CSV |
