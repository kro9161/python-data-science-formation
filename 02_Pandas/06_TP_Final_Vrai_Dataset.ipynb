## 🧩 Exercice 3 – Analyse : qui a survécu ?

Le dataset est propre. On peut maintenant répondre aux vraies questions.

*(Repartez du DataFrame nettoyé avec colonnes renommées)*

1. Quel est le **taux de survie global** ? (moyenne de `Survécu`)
2. Quel est le taux de survie **par classe** (`groupby` + `mean`) ?
3. Quel est le taux de survie **par sexe** ?
4. Quel est le taux de survie **par port d'embarquement** ?
5. Quelle est la **moyenne d'âge** des survivants vs des non-survivants ?
6. Combien de passagers par classe ? (`value_counts`)
7. Filtrez uniquement les **femmes de 1ère classe** — quel est leur taux de survie ?

````python
import pandas as pd

# Rechargement + nettoyage complet (pour que la cellule soit autonome)
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.rename(columns={
    "Survived":"Survécu", "Pclass":"Classe", "Sex":"Sexe",
    "Age":"Âge", "Fare":"Prix_billet", "Embarked":"Port"
})

# TODO 1 : taux de survie global
taux = df["Survécu"]....
print(f"1. Taux de survie global : {taux:.1%}")

# TODO 2 : taux de survie par classe
print("\n2. Survie par classe :")
print(df.groupby(...)[...].mean().round(3).apply(lambda x: f"{x:.1%}"))

# TODO 3 : taux de survie par sexe
print("\n3. Survie par sexe :")
print(df.groupby(...)[...].mean().round(3).apply(lambda x: f"{x:.1%}"))

# TODO 4 : taux de survie par port
print("\n4. Survie par port :")
print(df.groupby(...)[...].mean().round(3).apply(lambda x: f"{x:.1%}"))

# TODO 5 : âge moyen survivants vs non-survivants
print("\n5. Âge moyen :")
print(df.groupby(...)[...].mean().round(1))

# TODO 6 : nombre de passagers par classe
print("\n6. Passagers par classe :")
print(df[...].value_counts().sort_index())

# TODO 7 : femmes de 1ère classe
femmes_1ere = df[(df["Sexe"] == ...) & (df["Classe"] == ...)]
print(f"\n7. Femmes 1ère classe : {len(femmes_1ere)} passagères")
print(f"   Taux de survie     : {femmes_1ere['Survécu'].mean():.1%}")
````

## 🧩 Exercice 4 – Visualisation des résultats

Visualisez les analyses de l'exercice 3 sous forme de graphiques.

1. **Barres** : taux de survie par classe (horizontal)
2. **Barres** : taux de survie par sexe
3. **Histogramme** : distribution des âges de tous les passagers
4. **Camembert** : répartition des passagers par classe
5. **Grille 2×2** : les 4 graphiques ensemble

````python
import pandas as pd
import matplotlib.pyplot as plt

# Rechargement + nettoyage
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.rename(columns={
    "Survived":"Survécu", "Pclass":"Classe", "Sex":"Sexe",
    "Age":"Âge", "Fare":"Prix_billet", "Embarked":"Port"
})

# TODO 1 : barres horizontales taux de survie par classe
df.groupby("Classe")["Survécu"].mean().plot(
    kind=..., title="Taux de survie par classe",
    xlabel="Taux", color="steelblue", figsize=(7, 3)
)
plt.tight_layout(); plt.show()

# TODO 2 : barres taux de survie par sexe
df.groupby("Sexe")["Survécu"].mean().plot(
    kind=..., title="Taux de survie par sexe",
    color=["coral", "steelblue"], figsize=(6, 4)
)
plt.tight_layout(); plt.show()

# TODO 3 : histogramme des âges
df["Âge"].plot(
    kind=..., bins=...,
    title="Distribution des âges des passagers",
    xlabel="Âge", edgecolor="white", color="steelblue", figsize=(8, 4)
)
plt.tight_layout(); plt.show()

# TODO 4 : camembert répartition par classe
df["Classe"].value_counts().sort_index().plot(
    kind=..., autopct=..., ylabel=...,
    title="Répartition des passagers par classe",
    figsize=(5, 5)
)
plt.tight_layout(); plt.show()

# TODO 5 : grille 2×2
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

df.groupby("Classe")["Survécu"].mean().plot(kind="barh", ax=axes[0,0], title="Survie par classe",   color="steelblue")
df.groupby("Sexe")["Survécu"].mean().plot(  kind="bar",  ax=axes[0,1], title="Survie par sexe",    color=["coral","steelblue"])
df["Âge"].plot(                              kind="hist", ax=axes[1,0], title="Distribution des âges", bins=20, edgecolor="white")
df["Classe"].value_counts().sort_index().plot(kind="pie", ax=axes[1,1], title="Passagers par classe",  autopct="%1.0f%%", ylabel="")

plt.suptitle("Tableau de bord — Titanic", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
````

## 🧩 Exercice 5 – Export des résultats

Produisez un rapport Excel avec **3 feuilles** :

1. Feuille `"Données_propres"` : le DataFrame nettoyé et renommé
2. Feuille `"Analyse_survie"` : un tableau avec le taux de survie par classe ET par sexe (indice : `groupby(["Classe", "Sexe"])`)
3. Feuille `"Résumé"` : un petit DataFrame résumant les 3 stats principales :
   - Taux de survie global
   - Âge moyen
   - Prix moyen du billet

Exportez le tout dans `"rapport_titanic.xlsx"`.

````python
import pandas as pd

# Rechargement + nettoyage
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.rename(columns={
    "Survived":"Survécu", "Pclass":"Classe", "Sex":"Sexe",
    "Age":"Âge", "Fare":"Prix_billet", "Embarked":"Port"
})

# TODO : créer les 3 DataFrames à exporter

# Feuille 2 : analyse survie par classe ET sexe
analyse_survie = df.groupby([..., ...])["Survécu"].mean().round(3).reset_index()
analyse_survie.columns = ["Classe", "Sexe", "Taux_survie"]
print("Analyse survie :")
print(analyse_survie)

# Feuille 3 : résumé
resume = pd.DataFrame({
    "Indicateur" : ["Taux de survie global", "Âge moyen", "Prix moyen billet"],
    "Valeur"     : [
        f"{df['Survécu'].mean():.1%}",
        f"{df['Âge'].mean():.1f} ans",
        f"{df['Prix_billet'].mean():.2f} £"
    ]
})
print("\nRésumé :")
print(resume)

# TODO : écrire les 3 feuilles dans rapport_titanic.xlsx
with pd.ExcelWriter("rapport_titanic.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer,            sheet_name=..., index=False)
    analyse_survie.to_excel(writer, sheet_name=..., index=False)
    resume.to_excel(writer,         sheet_name=..., index=False)

print("\n✅ rapport_titanic.xlsx créé")
print("Feuilles :", pd.ExcelFile("rapport_titanic.xlsx").sheet_names)
````

Vous venez de réaliser une **analyse de données complète** sur un vrai dataset :

| Étape | Ce qu'on a fait | Compétences mobilisées |
|-------|----------------|------------------------|
| **Chargement** | `pd.read_csv(URL)` | Lire depuis internet ou fichier local |
| **Exploration** | `info()`, `describe()`, `isna()` | Comprendre la structure avant d'agir |
| **Nettoyage** | `drop()`, `fillna()`, `rename()` | Préparer les données pour l'analyse |
| **Analyse** | `groupby()`, `mean()`, filtres `&` | Répondre aux questions métier |
| **Visualisation** | `df.plot()`, grille `subplots` | Communiquer les résultats |
| **Export** | `ExcelWriter` multi-feuilles | Livrer le travail |

**Ce pipeline — charger → explorer → nettoyer → analyser → visualiser → exporter — c'est exactement ce qu'on fait dans un vrai projet data en entreprise.**

La seule différence avec un projet réel : les données viennent d'une base SQL, d'une API ou d'un ERP au lieu d'un fichier CSV. Mais la logique Pandas reste exactement la même.
